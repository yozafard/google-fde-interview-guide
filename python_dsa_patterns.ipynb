```python
"""
GOOGLE INTERVIEW: PYTHON DSA TEMPLATES & PATTERNS
==================================================
Forward Deployed Engineer - Applied AI

MEMORIZE THESE PATTERNS - They cover 80% of interview questions.
Practice typing these WITHOUT an IDE until they're muscle memory.
"""

# =============================================================================
# SECTION 1: ESSENTIAL DATA STRUCTURE OPERATIONS
# =============================================================================

# --- Hash Table / Dictionary ---
# Time: O(1) average for get/set/delete
# Use for: Counting, caching, quick lookups, finding pairs

def two_sum(nums: list[int], target: int) -> list[int]:
    """Classic two-sum using hash map. O(n) time, O(n) space."""
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

def count_frequency(arr: list) -> dict:
    """Count occurrences of each element."""
    from collections import Counter
    return Counter(arr)  # Or manually: freq = {}; freq[x] = freq.get(x, 0) + 1


# --- Arrays & Strings ---

def reverse_string_in_place(s: list[str]) -> None:
    """Reverse list of chars in-place. O(n) time, O(1) space."""
    left, right = 0, len(s) - 1
    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1


# =============================================================================
# SECTION 2: TWO POINTERS PATTERN
# =============================================================================
# Use when: Sorted arrays, finding pairs, palindromes, removing duplicates

def is_palindrome(s: str) -> bool:
    """Check if string is palindrome (alphanumeric only)."""
    left, right = 0, len(s) - 1
    while left < right:
        # Skip non-alphanumeric
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True

def remove_duplicates_sorted(nums: list[int]) -> int:
    """Remove duplicates in-place from sorted array. Return new length."""
    if not nums:
        return 0
    write_idx = 1
    for read_idx in range(1, len(nums)):
        if nums[read_idx] != nums[read_idx - 1]:
            nums[write_idx] = nums[read_idx]
            write_idx += 1
    return write_idx

def three_sum(nums: list[int]) -> list[list[int]]:
    """Find all unique triplets that sum to zero. O(n²) time."""
    nums.sort()
    result = []
    for i in range(len(nums) - 2):
        # Skip duplicates for first element
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                result.append([nums[i], nums[left], nums[right]])
                # Skip duplicates
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
    return result


# =============================================================================
# SECTION 3: SLIDING WINDOW PATTERN
# =============================================================================
# Use when: Contiguous subarrays/substrings, max/min in window, fixed or variable size

def max_sum_subarray_fixed(arr: list[int], k: int) -> int:
    """Maximum sum of any subarray of size k. O(n) time."""
    if len(arr) < k:
        return 0

    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]  # Slide: add new, remove old
        max_sum = max(max_sum, window_sum)

    return max_sum

def longest_substring_k_distinct(s: str, k: int) -> int:
    """Longest substring with at most k distinct characters. O(n) time."""
    from collections import defaultdict

    char_count = defaultdict(int)
    left = 0
    max_length = 0

    for right in range(len(s)):
        char_count[s[right]] += 1

        # Shrink window while we have more than k distinct chars
        while len(char_count) > k:
            char_count[s[left]] -= 1
            if char_count[s[left]] == 0:
                del char_count[s[left]]
            left += 1

        max_length = max(max_length, right - left + 1)

    return max_length

def min_window_substring(s: str, t: str) -> str:
    """Minimum window in s containing all chars of t. O(n) time."""
    from collections import Counter

    if not s or not t:
        return ""

    need = Counter(t)
    have = {}
    required = len(need)
    formed = 0

    left = 0
    min_len = float('inf')
    result = ""

    for right in range(len(s)):
        char = s[right]
        have[char] = have.get(char, 0) + 1

        if char in need and have[char] == need[char]:
            formed += 1

        # Try to contract window
        while formed == required:
            if right - left + 1 < min_len:
                min_len = right - left + 1
                result = s[left:right + 1]

            left_char = s[left]
            have[left_char] -= 1
            if left_char in need and have[left_char] < need[left_char]:
                formed -= 1
            left += 1

    return result


# =============================================================================
# SECTION 4: BINARY SEARCH PATTERNS
# =============================================================================
# Use when: Sorted array, finding target/boundary, minimizing/maximizing something

def binary_search_standard(nums: list[int], target: int) -> int:
    """Standard binary search. Returns index or -1. O(log n) time."""
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2  # Avoid overflow
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

def binary_search_leftmost(nums: list[int], target: int) -> int:
    """Find leftmost (first) occurrence of target."""
    left, right = 0, len(nums)

    while left < right:
        mid = left + (right - left) // 2
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid

    return left if left < len(nums) and nums[left] == target else -1

def binary_search_rightmost(nums: list[int], target: int) -> int:
    """Find rightmost (last) occurrence of target."""
    left, right = 0, len(nums)

    while left < right:
        mid = left + (right - left) // 2
        if nums[mid] <= target:
            left = mid + 1
        else:
            right = mid

    return left - 1 if left > 0 and nums[left - 1] == target else -1

def search_rotated_array(nums: list[int], target: int) -> int:
    """Search in rotated sorted array. O(log n) time."""
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid

        # Check which half is sorted
        if nums[left] <= nums[mid]:  # Left half is sorted
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:  # Right half is sorted
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1


# =============================================================================
# SECTION 5: LINKED LIST PATTERNS
# =============================================================================

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse_linked_list(head: ListNode) -> ListNode:
    """Reverse a linked list iteratively. O(n) time, O(1) space."""
    prev = None
    curr = head

    while curr:
        next_temp = curr.next  # Save next
        curr.next = prev       # Reverse pointer
        prev = curr            # Move prev forward
        curr = next_temp       # Move curr forward

    return prev

def has_cycle(head: ListNode) -> bool:
    """Detect cycle using Floyd's algorithm. O(n) time, O(1) space."""
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True

    return False

def find_cycle_start(head: ListNode) -> ListNode:
    """Find where cycle begins. Returns None if no cycle."""
    slow = fast = head

    # Phase 1: Detect cycle
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return None

    # Phase 2: Find start
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next

    return slow

def merge_two_sorted_lists(l1: ListNode, l2: ListNode) -> ListNode:
    """Merge two sorted linked lists. O(n+m) time."""
    dummy = ListNode(0)
    tail = dummy

    while l1 and l2:
        if l1.val <= l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    tail.next = l1 or l2
    return dummy.next

def find_middle(head: ListNode) -> ListNode:
    """Find middle node (slow/fast pointer). Returns second middle if even."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow


# =============================================================================
# SECTION 6: TREE PATTERNS
# =============================================================================

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# --- DFS Traversals (Recursive) ---

def inorder_recursive(root: TreeNode) -> list[int]:
    """Left -> Root -> Right. For BST: gives sorted order."""
    result = []
    def dfs(node):
        if node:
            dfs(node.left)
            result.append(node.val)
            dfs(node.right)
    dfs(root)
    return result

def preorder_recursive(root: TreeNode) -> list[int]:
    """Root -> Left -> Right. Good for copying trees."""
    result = []
    def dfs(node):
        if node:
            result.append(node.val)
            dfs(node.left)
            dfs(node.right)
    dfs(root)
    return result

def postorder_recursive(root: TreeNode) -> list[int]:
    """Left -> Right -> Root. Good for deleting trees."""
    result = []
    def dfs(node):
        if node:
            dfs(node.left)
            dfs(node.right)
            result.append(node.val)
    dfs(root)
    return result

# --- DFS Traversals (Iterative) ---

def inorder_iterative(root: TreeNode) -> list[int]:
    """Iterative inorder using stack."""
    result = []
    stack = []
    curr = root

    while stack or curr:
        while curr:
            stack.append(curr)
            curr = curr.left
        curr = stack.pop()
        result.append(curr.val)
        curr = curr.right

    return result

# --- BFS Traversal ---

def level_order(root: TreeNode) -> list[list[int]]:
    """Level-order traversal (BFS). O(n) time."""
    from collections import deque

    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        level_size = len(queue)
        level = []
        for _ in range(level_size):
            node = queue.popleft()
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(level)

    return result

# --- Common Tree Problems ---

def max_depth(root: TreeNode) -> int:
    """Maximum depth of binary tree."""
    if not root:
        return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))

def is_balanced(root: TreeNode) -> bool:
    """Check if tree is height-balanced."""
    def check(node):
        if not node:
            return 0
        left = check(node.left)
        right = check(node.right)
        if left == -1 or right == -1 or abs(left - right) > 1:
            return -1
        return 1 + max(left, right)

    return check(root) != -1

def is_valid_bst(root: TreeNode) -> bool:
    """Validate binary search tree."""
    def validate(node, min_val, max_val):
        if not node:
            return True
        if node.val <= min_val or node.val >= max_val:
            return False
        return (validate(node.left, min_val, node.val) and
                validate(node.right, node.val, max_val))

    return validate(root, float('-inf'), float('inf'))

def lowest_common_ancestor(root: TreeNode, p: TreeNode, q: TreeNode) -> TreeNode:
    """Find LCA of two nodes in binary tree."""
    if not root or root == p or root == q:
        return root

    left = lowest_common_ancestor(root.left, p, q)
    right = lowest_common_ancestor(root.right, p, q)

    if left and right:
        return root
    return left or right


# =============================================================================
# SECTION 7: GRAPH PATTERNS
# =============================================================================

# --- Graph Representations ---

def build_adjacency_list(edges: list[list[int]], directed=False) -> dict:
    """Build adjacency list from edge list."""
    from collections import defaultdict
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
        if not directed:
            graph[v].append(u)
    return graph

# --- BFS (Shortest Path in Unweighted Graph) ---

def bfs_shortest_path(graph: dict, start: int, end: int) -> int:
    """Find shortest path length using BFS. Returns -1 if no path."""
    from collections import deque

    if start == end:
        return 0

    visited = {start}
    queue = deque([(start, 0)])

    while queue:
        node, dist = queue.popleft()
        for neighbor in graph[node]:
            if neighbor == end:
                return dist + 1
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, dist + 1))

    return -1

# --- DFS (Connected Components, Cycle Detection) ---

def count_connected_components(n: int, edges: list[list[int]]) -> int:
    """Count connected components in undirected graph."""
    graph = build_adjacency_list(edges)
    visited = set()
    components = 0

    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor)

    for node in range(n):
        if node not in visited:
            dfs(node)
            components += 1

    return components

def has_cycle_directed(graph: dict, n: int) -> bool:
    """Detect cycle in directed graph using colors."""
    WHITE, GRAY, BLACK = 0, 1, 2
    color = {i: WHITE for i in range(n)}

    def dfs(node):
        color[node] = GRAY
        for neighbor in graph.get(node, []):
            if color[neighbor] == GRAY:  # Back edge = cycle
                return True
            if color[neighbor] == WHITE and dfs(neighbor):
                return True
        color[node] = BLACK
        return False

    for node in range(n):
        if color[node] == WHITE and dfs(node):
            return True
    return False

# --- Topological Sort (Kahn's Algorithm) ---

def topological_sort(n: int, edges: list[list[int]]) -> list[int]:
    """Topological sort using BFS (Kahn's). Returns [] if cycle exists."""
    from collections import deque, defaultdict

    graph = defaultdict(list)
    in_degree = [0] * n

    for u, v in edges:
        graph[u].append(v)
        in_degree[v] += 1

    queue = deque([i for i in range(n) if in_degree[i] == 0])
    result = []

    while queue:
        node = queue.popleft()
        result.append(node)
        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    return result if len(result) == n else []  # Cycle if not all nodes


# =============================================================================
# SECTION 8: HEAP / PRIORITY QUEUE PATTERNS
# =============================================================================

import heapq

def find_kth_largest(nums: list[int], k: int) -> int:
    """Find kth largest element. O(n log k) time."""
    # Use min-heap of size k
    heap = nums[:k]
    heapq.heapify(heap)

    for num in nums[k:]:
        if num > heap[0]:
            heapq.heapreplace(heap, num)  # Pop smallest, push new

    return heap[0]

def merge_k_sorted_lists(lists: list[ListNode]) -> ListNode:
    """Merge k sorted linked lists. O(n log k) time."""
    heap = []

    # Add first node of each list
    for i, lst in enumerate(lists):
        if lst:
            heapq.heappush(heap, (lst.val, i, lst))

    dummy = ListNode(0)
    tail = dummy

    while heap:
        val, i, node = heapq.heappop(heap)
        tail.next = node
        tail = tail.next
        if node.next:
            heapq.heappush(heap, (node.next.val, i, node.next))

    return dummy.next

def top_k_frequent(nums: list[int], k: int) -> list[int]:
    """Find k most frequent elements. O(n log k) time."""
    from collections import Counter

    count = Counter(nums)
    return heapq.nlargest(k, count.keys(), key=count.get)


# =============================================================================
# SECTION 9: DYNAMIC PROGRAMMING PATTERNS
# =============================================================================

# --- 1D DP ---

def climb_stairs(n: int) -> int:
    """Number of ways to climb n stairs (1 or 2 steps at a time)."""
    if n <= 2:
        return n

    prev2, prev1 = 1, 2
    for _ in range(3, n + 1):
        curr = prev1 + prev2
        prev2, prev1 = prev1, curr

    return prev1

def house_robber(nums: list[int]) -> int:
    """Max money without robbing adjacent houses."""
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0]

    prev2, prev1 = nums[0], max(nums[0], nums[1])
    for i in range(2, len(nums)):
        curr = max(prev1, prev2 + nums[i])
        prev2, prev1 = prev1, curr

    return prev1

def longest_increasing_subsequence(nums: list[int]) -> int:
    """Length of LIS. O(n log n) using binary search."""
    from bisect import bisect_left

    tails = []
    for num in nums:
        pos = bisect_left(tails, num)
        if pos == len(tails):
            tails.append(num)
        else:
            tails[pos] = num

    return len(tails)

# --- 2D DP ---

def unique_paths(m: int, n: int) -> int:
    """Number of unique paths in m x n grid (only right/down)."""
    dp = [1] * n
    for _ in range(1, m):
        for j in range(1, n):
            dp[j] += dp[j - 1]
    return dp[-1]

def longest_common_subsequence(text1: str, text2: str) -> int:
    """Length of LCS of two strings. O(m*n) time and space."""
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

def coin_change(coins: list[int], amount: int) -> int:
    """Minimum coins needed. Returns -1 if impossible."""
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1


# =============================================================================
# SECTION 10: SORTING ALGORITHMS (Know These Cold!)
# =============================================================================

def merge_sort(arr: list[int]) -> list[int]:
    """Merge sort. O(n log n) time, O(n) space. STABLE."""
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left: list[int], right: list[int]) -> list[int]:
    """Merge two sorted arrays."""
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort(arr: list[int], low: int = 0, high: int = None) -> None:
    """Quick sort in-place. O(n log n) average, O(n²) worst. NOT STABLE."""
    if high is None:
        high = len(arr) - 1

    if low < high:
        pivot_idx = partition(arr, low, high)
        quick_sort(arr, low, pivot_idx - 1)
        quick_sort(arr, pivot_idx + 1, high)

def partition(arr: list[int], low: int, high: int) -> int:
    """Lomuto partition scheme."""
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


# =============================================================================
# SECTION 11: TRIE (Prefix Tree)
# =============================================================================

class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    """Prefix tree for efficient string operations."""

    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        """Insert word. O(m) time where m is word length."""
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word: str) -> bool:
        """Check if exact word exists."""
        node = self._find_node(word)
        return node is not None and node.is_end

    def starts_with(self, prefix: str) -> bool:
        """Check if any word starts with prefix."""
        return self._find_node(prefix) is not None

    def _find_node(self, prefix: str) -> TrieNode:
        node = self.root
        for char in prefix:
            if char not in node.children:
                return None
            node = node.children[char]
        return node


# =============================================================================
# SECTION 12: BIG-O CHEAT SHEET
# =============================================================================
"""
TIME COMPLEXITY RANKINGS (Best to Worst):
O(1)        - Constant (hash lookup, array index)
O(log n)    - Logarithmic (binary search)
O(n)        - Linear (single loop, linear search)
O(n log n)  - Linearithmic (merge sort, heap sort)
O(n²)       - Quadratic (nested loops, bubble sort)
O(2^n)      - Exponential (recursive fibonacci, subsets)
O(n!)       - Factorial (permutations)

SPACE COMPLEXITY:
- In-place algorithms: O(1) extra space
- Recursive algorithms: O(depth) for call stack
- Hash maps/sets: O(n) typically
- 2D DP: O(m*n), can often optimize to O(n)

OPERATION COMPLEXITIES:
Array/List:
  - Access by index: O(1)
  - Search: O(n)
  - Insert/Delete at end: O(1) amortized
  - Insert/Delete at position: O(n)

Hash Table (dict/set):
  - Insert/Delete/Search: O(1) average, O(n) worst

Heap:
  - Insert: O(log n)
  - Extract min/max: O(log n)
  - Peek: O(1)
  - Heapify: O(n)

BST (balanced):
  - Search/Insert/Delete: O(log n)
  - Unbalanced worst case: O(n)

SORTING:
  - Comparison-based lower bound: O(n log n)
  - Counting/Radix sort: O(n) for limited range
  - Python's sort: Timsort, O(n log n)
"""


# =============================================================================
# SECTION 13: INTERVIEW COMMUNICATION TEMPLATE
# =============================================================================
"""
STEP 1: CLARIFY (2-3 minutes)
------------------------------
"Before I start, let me make sure I understand the problem correctly..."
- What are the input constraints? (size, range, types)
- Are there edge cases? (empty input, single element, negatives)
- What should I return? (index, value, list, boolean)
- Is the input sorted? Can I modify it?
- Are there duplicates? How to handle them?

STEP 2: EXAMPLES (1-2 minutes)
------------------------------
"Let me walk through an example to verify my understanding..."
- Use the given example first
- Create a simple edge case
- Create a more complex case if needed

STEP 3: APPROACH (2-3 minutes)
------------------------------
"I'm thinking of a few approaches..."
- Start with brute force (acknowledge it, state complexity)
- "A brute force approach would be O(n²) by checking every pair..."
- "But we can do better using [technique] to get O(n)..."
- State the optimal approach and WHY it works

STEP 4: CODE (15-20 minutes)
----------------------------
"I'll start coding now. I'll talk through my logic as I go..."
- Write clean, readable code
- Use meaningful variable names
- Add brief comments for complex logic
- Handle edge cases explicitly

STEP 5: TEST (3-5 minutes)
--------------------------
"Let me trace through with an example..."
- Walk through your code line by line
- Use a simple example first
- Then test an edge case
- Fix any bugs you find (this is expected!)

STEP 6: COMPLEXITY (1 minute)
-----------------------------
"For complexity analysis..."
- Time: "We iterate through n elements once, so O(n)"
- Space: "We use a hash map that could store n elements, so O(n)"

MAGIC PHRASES:
- "That's a great question, let me think about that..."
- "I'm going to think out loud here..."
- "I notice that..." (when you spot a pattern)
- "One optimization we could make is..."
- "Let me reconsider this approach..."
"""
```
