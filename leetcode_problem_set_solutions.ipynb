### Two Sum

**Difficulty:** Easy  
**Pattern:** Hash Map  
**Reference:** LeetCode 1  
**Solution to study:** `two_sum`

**What the question is testing:** Whether you can recognize `Hash Map` and maintain the right invariant while handling edge cases.

**Best approach:** Use a dictionary to turn repeated lookup into average O(1) work.

**How to recognize it:** Look for pair lookup, counting, grouping, or seen-before checks.

**What to say in an interview:** Start with the brute force idea, then explain why `Hash Map` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Create a `seen` dictionary where each number points to its index.
2. For every number, compute the complement `target - x`.
3. If the complement is already in `seen`, the pair is found; otherwise store the current number.

**Edge cases to check:** duplicates, empty input, repeated values.


```python
# Step-by-step walkthrough:
# Step 1: Create a `seen` dictionary where each number points to its index.
# Step 2: For every number, compute the complement `target - x`.
# Step 3: If the complement is already in `seen`, the pair is found; otherwise store the
#         current number.

def two_sum(nums, target):
    # Hash map. Time O(n), space O(n).
    seen = {}
    for i, x in enumerate(nums):
        need = target - x
        if need in seen:
            return [seen[need], i]
        seen[x] = i
    return []
```

### Contains Duplicate

**Difficulty:** Easy  
**Pattern:** Hash Set  
**Reference:** LeetCode 217  
**Solution to study:** `contains_duplicate`

**What the question is testing:** Whether you can recognize `Hash Set` and maintain the right invariant while handling edge cases.

**Best approach:** Use a set when you only need to know whether a value exists.

**How to recognize it:** Look for duplicate detection or membership questions.

**What to say in an interview:** Start with the brute force idea, then explain why `Hash Set` removes wasted work. State the invariant before coding.

**How the solution works:**

2. If the set is shorter than the original list, at least one duplicate existed.

**Edge cases to check:** empty input, one item, all same values.


```python
# Step-by-step walkthrough:
# Step 1: Convert the list to a set, which removes duplicates.
# Step 2: If the set is shorter than the original list, at least one duplicate existed.

def contains_duplicate(nums):
    # Hash set. Time O(n), space O(n).
    return len(nums) != len(set(nums))
```

### Group Anagrams

**Difficulty:** Medium  
**Pattern:** Hash Map  
**Reference:** LeetCode 49  
**Solution to study:** `group_anagrams`

**What the question is testing:** Whether you can recognize `Hash Map` and maintain the right invariant while handling edge cases.

**Best approach:** Use a dictionary to turn repeated lookup into average O(1) work.

**How to recognize it:** Look for pair lookup, counting, grouping, or seen-before checks.

**What to say in an interview:** Start with the brute force idea, then explain why `Hash Map` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Sort each word to create a shared key for anagrams.
2. Append each word to the list for that key in a dictionary.
3. Return the grouped dictionary values.

**Edge cases to check:** duplicates, empty input, repeated values.


```python
# Step-by-step walkthrough:
# Step 1: Sort each word to create a shared key for anagrams.
# Step 2: Append each word to the list for that key in a dictionary.
# Step 3: Return the grouped dictionary values.

def group_anagrams(strs):
    # Canonical key by sorted characters. Time O(n * k log k), space O(n * k).
    groups = defaultdict(list)
    for word in strs:
        groups["".join(sorted(word))].append(word)
    return list(groups.values())
```

**What to say in an interview:** Start with the brute force idea, then explain why `Hash Map + Heap` removes wasted work. State the invariant before coding.

**How the solution works:**

2. Use `heapq.nlargest` to select the keys with the highest counts.

```python
# Step-by-step walkthrough:
# Step 1: Count each number with `Counter`.
# Step 2: Use `heapq.nlargest` to select the keys with the highest counts.
# Step 3: This avoids manually sorting by frequency in the solution logic.

def top_k_frequent(nums, k):
    # Counter + heap helper. Time O(n log k), space O(n).
    count = Counter(nums)
    return heapq.nlargest(k, count.keys(), key=count.get)
```

### Product of Array Except Self

**How to recognize it:** Match the problem wording to the pattern column.

**What to say in an interview:** Start with the brute force idea, then explain why `Prefix/Suffix` removes wasted work. State the invariant before coding.

**How the solution works:**

1. First pass stores the product of everything to the left of each index.
2. Second pass multiplies by the product of everything to the right.
3. No division is needed, so zero values are handled correctly.

**Edge cases to check:** empty input, one element, duplicate values.


```python
# Step-by-step walkthrough:
# Step 1: First pass stores the product of everything to the left of each index.
# Step 2: Second pass multiplies by the product of everything to the right.
# Step 3: No division is needed, so zero values are handled correctly.

def product_except_self(nums):
    # Prefix/suffix without division. Time O(n), space O(1) extra besides output.
    res = [1] * len(nums)
    prefix = 1
    for i, x in enumerate(nums):
        res[i] = prefix
        prefix *= x
    suffix = 1
    for i in range(len(nums) - 1, -1, -1):
        res[i] *= suffix
        suffix *= nums[i]
    return res
```

### Valid Palindrome

**Difficulty:** Easy  
**Pattern:** Two Pointers  
**Reference:** LeetCode 125  
**Solution to study:** `is_palindrome`

**What the question is testing:** Whether you can recognize `Two Pointers` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Place one pointer at the start and one at the end.
2. Skip characters that are not letters or digits.
3. Compare lowercase characters, then move both pointers inward.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Place one pointer at the start and one at the end.
# Step 2: Skip characters that are not letters or digits.
# Step 3: Compare lowercase characters, then move both pointers inward.

def is_palindrome(s):
    # Opposite-end pointers. Time O(n), space O(1).
    left, right = 0, len(s) - 1
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True
```

### Two Sum II (sorted)

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 167  
**Solution to study:** `two_sum_sorted`

**What the question is testing:** Whether you can recognize `Two Pointers` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Use the sorted order to avoid a hash map.
2. If the two-pointer sum is too small, move `left` rightward.
3. If it is too large, move `right` leftward.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Use the sorted order to avoid a hash map.
# Step 2: If the two-pointer sum is too small, move `left` rightward.
# Step 3: If it is too large, move `right` leftward.

def two_sum_sorted(numbers, target):
    # Sorted array lets us move the pointer that is too small or too large.
    # Time O(n), space O(1). Returns 1-indexed positions.
    left, right = 0, len(numbers) - 1
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]
        if total < target:
            left += 1
        else:
            right -= 1
    return []
```

### 3Sum

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 15  
**Solution to study:** `three_sum`

**What the question is testing:** Whether you can recognize `Two Pointers` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Sort first so duplicates can be skipped and two pointers can be used.
2. Fix one number at index `i`, then search for the other two numbers with `left` and `right`.
3. After finding a triplet, skip duplicate left and right values.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Sort first so duplicates can be skipped and two pointers can be used.
# Step 2: Fix one number at index `i`, then search for the other two numbers with `left`
#         and `right`.
# Step 3: After finding a triplet, skip duplicate left and right values.

def three_sum(nums):
    # Sort + two pointers. Time O(n^2), space O(1) extra besides output.
    nums.sort()
    res = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            elif total < 0:
                left += 1
            else:
                right -= 1
    return res
```

### Container With Most Water

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 11  
**Solution to study:** `max_area`

**What the question is testing:** Whether you can recognize `Two Pointers` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Start with the widest possible container.
2. Compute the area using the shorter wall.
3. Move the shorter wall inward because moving the taller wall cannot improve the limiting height.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Start with the widest possible container.
# Step 2: Compute the area using the shorter wall.
# Step 3: Move the shorter wall inward because moving the taller wall cannot improve the
#         limiting height.

def max_area(height):
    # Move the shorter wall because width always shrinks. Time O(n), space O(1).
    left, right = 0, len(height) - 1
    best = 0
    while left < right:
        best = max(best, min(height[left], height[right]) * (right - left))
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    return best
```

### Trapping Rain Water

**Difficulty:** Hard  
**Pattern:** Two Pointers/DP  
**Reference:** LeetCode 42  
**Solution to study:** `trap`

**What the question is testing:** Whether you can recognize `Two Pointers/DP` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers/DP` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Track the best wall seen from the left and from the right.
2. Move the side with the smaller current height because that side limits trapped water.
3. Add the difference between the side max and current height.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Track the best wall seen from the left and from the right.
# Step 2: Move the side with the smaller current height because that side limits trapped
#         water.
# Step 3: Add the difference between the side max and current height.

def trap(height):
    # Two pointers with left/right max. Time O(n), space O(1).
    left, right = 0, len(height) - 1
    left_max = right_max = water = 0
    while left < right:
        if height[left] < height[right]:
            left_max = max(left_max, height[left])
            water += left_max - height[left]
            left += 1
        else:
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1
    return water
```

### Remove Duplicates from Sorted Array

**Difficulty:** Easy  
**Pattern:** Two Pointers  
**Reference:** LeetCode 26  
**Solution to study:** `remove_duplicates`

**What the question is testing:** Whether you can recognize `Two Pointers` and maintain the right invariant while handling edge cases.

**Best approach:** Use left/right or read/write pointers to avoid nested loops.

**How to recognize it:** Look for sorted arrays, palindromes, pairs, in-place edits, or boundary comparisons.

**What to say in an interview:** Start with the brute force idea, then explain why `Two Pointers` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Use `read` to scan the array and `write` to place unique values.
2. Because the array is sorted, a new value differs from the previous value.
3. Return `write`, the length of the unique prefix.

**Edge cases to check:** duplicates, empty input, pointers crossing.


```python
# Step-by-step walkthrough:
# Step 1: Use `read` to scan the array and `write` to place unique values.
# Step 2: Because the array is sorted, a new value differs from the previous value.
# Step 3: Return `write`, the length of the unique prefix.

def remove_duplicates(nums):
    # Write pointer keeps the next unique position. Time O(n), space O(1).
    if not nums:
        return 0
    write = 1
    for read in range(1, len(nums)):
        if nums[read] != nums[read - 1]:
            nums[write] = nums[read]
            write += 1
    return write
```

### Best Time to Buy and Sell Stock

**Difficulty:** Easy  
**Pattern:** Sliding Window  
**Reference:** LeetCode 121  
**Solution to study:** `max_profit`

**What the question is testing:** Whether you can recognize `Sliding Window` and maintain the right invariant while handling edge cases.

**Best approach:** Maintain a contiguous window and expand or shrink it as constraints change.

**How to recognize it:** Look for substring/subarray, longest, shortest, fixed k, or at most k.

**What to say in an interview:** Start with the brute force idea, then explain why `Sliding Window` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Track the lowest buy price seen so far.
2. For each price, compute the profit if sold today.
3. Keep the maximum profit across all days.

**Edge cases to check:** empty string, k is 0, repeated characters.


```python
# Step-by-step walkthrough:
# Step 1: Track the lowest buy price seen so far.
# Step 2: For each price, compute the profit if sold today.
# Step 3: Keep the maximum profit across all days.

def max_profit(prices):
    # Track lowest buy so far. Time O(n), space O(1).
    best = 0
    buy = float("inf")
    for price in prices:
        buy = min(buy, price)
        best = max(best, price - buy)
    return best
```

### Maximum Subarray

**Difficulty:** Medium  
**Pattern:** Kadane's  
**Reference:** LeetCode 53  
**Solution to study:** `max_sub_array`

**What the question is testing:** Whether you can recognize `Kadane's` and maintain the right invariant while handling edge cases.

**Best approach:** Track the best subarray ending here and the best seen globally.

**How to recognize it:** Look for maximum contiguous subarray sum.

**What to say in an interview:** Start with the brute force idea, then explain why `Kadane's` removes wasted work. State the invariant before coding.

**How the solution works:**

1. `curr` means the best subarray sum ending at the current index.
2. At each number, choose between starting fresh or extending the previous subarray.
3. `best` stores the best sum seen anywhere.

**Edge cases to check:** all negative numbers, one element.


```python
# Step-by-step walkthrough:
# Step 1: `curr` means the best subarray sum ending at the current index.
# Step 2: At each number, choose between starting fresh or extending the previous
#         subarray.
# Step 3: `best` stores the best sum seen anywhere.

def max_sub_array(nums):
    # Kadane's algorithm. Time O(n), space O(1).
    best = curr = nums[0]
    for x in nums[1:]:
        curr = max(x, curr + x)
        best = max(best, curr)
    return best
```

```python
# Step-by-step walkthrough:
# Step 1: Keep a window with no repeated characters.
# Step 2: Store each character?s last seen index.
# Step 3: When a repeat appears inside the current window, move `left` past the previous
#         occurrence.

def length_of_longest_substring(s):
    # Variable window with last-seen positions. Time O(n), space O(k).
    last = {}
    left = best = 0
    for right, ch in enumerate(s):
        if ch in last and last[ch] >= left:
            left = last[ch] + 1
        last[ch] = right
        best = max(best, right - left + 1)
    return best
```

### Longest Repeating Character Replacement

**Best approach:** Maintain a contiguous window and expand or shrink it as constraints change.

**How to recognize it:** Look for substring/subarray, longest, shortest, fixed k, or at most k.

**What to say in an interview:** Start with the brute force idea, then explain why `Sliding Window` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Track character counts inside the current window.
2. The window is valid when `window_size - most_common_count <= k`.
3. Shrink from the left only when more than `k` replacements are needed.

**Edge cases to check:** empty string, k is 0, repeated characters.


```python
# Step-by-step walkthrough:
# Step 1: Track character counts inside the current window.
# Step 2: The window is valid when `window_size - most_common_count <= k`.
# Step 3: Shrink from the left only when more than `k` replacements are needed.

def character_replacement(s, k):
    # Keep window where changes_needed <= k. Time O(n), space O(1).
    count = defaultdict(int)
    left = best = max_count = 0
    for right, ch in enumerate(s):
        count[ch] += 1
        max_count = max(max_count, count[ch])
        while (right - left + 1) - max_count > k:
            count[s[left]] -= 1
            left += 1
        best = max(best, right - left + 1)
    return best
```

### Minimum Window Substring

**Difficulty:** Hard  
**Pattern:** Sliding Window  
**Reference:** LeetCode 76  
**Solution to study:** `min_window`

**What the question is testing:** Whether you can recognize `Sliding Window` and maintain the right invariant while handling edge cases.

**Best approach:** Maintain a contiguous window and expand or shrink it as constraints change.

**How to recognize it:** Look for substring/subarray, longest, shortest, fixed k, or at most k.

**What to say in an interview:** Start with the brute force idea, then explain why `Sliding Window` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Count the characters required by `t`.
2. Expand right until the current window covers all required characters.
3. Then shrink from the left to make the valid window as small as possible.

**Edge cases to check:** empty string, k is 0, repeated characters.


```python
# Step-by-step walkthrough:
# Step 1: Count the characters required by `t`.
# Step 2: Expand right until the current window covers all required characters.
# Step 3: Then shrink from the left to make the valid window as small as possible.

def min_window(s, t):
    # Expand until valid, shrink to minimal. Time O(n), space O(k).
    need = Counter(t)
    missing = len(t)
    left = start = end = 0
    for right, ch in enumerate(s, 1):
        if need[ch] > 0:
            missing -= 1
        need[ch] -= 1
        if missing == 0:
            while left < right and need[s[left]] < 0:
                need[s[left]] += 1
                left += 1
            if end == 0 or right - left < end - start:
                start, end = left, right
            need[s[left]] += 1
            missing += 1
            left += 1
    return s[start:end]
```

### Sliding Window Maximum

**Difficulty:** Hard  
**Pattern:** Deque  
**Reference:** LeetCode 239  
**Solution to study:** `max_sliding_window`

**What the question is testing:** Whether you can recognize `Deque` and maintain the right invariant while handling edge cases.

**Best approach:** Use a monotonic deque to keep only useful window candidates.

**How to recognize it:** Look for sliding window maximum/minimum.

**What to say in an interview:** Start with the brute force idea, then explain why `Deque` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Store indexes in a deque, not values, so stale indexes can be removed.
2. Keep the deque decreasing by value, so the front is always the max.
3. Once a full window is formed, append `nums[q[0]]`.

**Edge cases to check:** stale indexes, duplicate values.


```python
# Step-by-step walkthrough:
# Step 1: Store indexes in a deque, not values, so stale indexes can be removed.
# Step 2: Keep the deque decreasing by value, so the front is always the max.
# Step 3: Once a full window is formed, append `nums[q[0]]`.

def max_sliding_window(nums, k):
    # Monotonic deque stores candidate indexes. Time O(n), space O(k).
    q = deque()
    res = []
    for i, x in enumerate(nums):
        while q and q[0] <= i - k:
            q.popleft()
        while q and nums[q[-1]] <= x:
            q.pop()
        q.append(i)
        if i >= k - 1:
            res.append(nums[q[0]])
    return res
```

### Valid Parentheses

**Difficulty:** Easy  
**Pattern:** Stack  
**Reference:** LeetCode 20  
**Solution to study:** `is_valid_parentheses`

**What the question is testing:** Whether you can recognize `Stack` and maintain the right invariant while handling edge cases.

**Best approach:** Use a stack when the most recent unresolved item matters.

**How to recognize it:** Look for matching brackets, postfix expressions, or next greater element.

**What to say in an interview:** Start with the brute force idea, then explain why `Stack` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Push opening brackets onto a stack.
2. When a closing bracket appears, the stack top must be its matching opener.
3. At the end, the stack must be empty.

**Edge cases to check:** empty stack, invalid order, duplicate heights.


```python
# Step-by-step walkthrough:
# Step 1: Push opening brackets onto a stack.
# Step 2: When a closing bracket appears, the stack top must be its matching opener.
# Step 3: At the end, the stack must be empty.

def is_valid_parentheses(s):
    # Stack matching. Time O(n), space O(n).
    pairs = {")": "(", "]": "[", "}": "{"}
    stack = []
    for ch in s:
        if ch in pairs.values():
            stack.append(ch)
        elif not stack or stack.pop() != pairs[ch]:
            return False
    return not stack
```

```python
# Step-by-step walkthrough:
# Step 1: Push numbers onto a stack.
# Step 2: For an operator, pop the last two numbers in operand order.
# Step 3: Apply the operation and push the result back.

def eval_rpn(tokens):
    # Stack operands. Time O(n), space O(n).
    stack = []
    for token in tokens:
        if token not in "+-*/":
            stack.append(int(token))
            continue
        b, a = stack.pop(), stack.pop()
        if token == "+":
            stack.append(a + b)
        elif token == "-":
            stack.append(a - b)
        elif token == "*":
            stack.append(a * b)
        else:
            stack.append(int(a / b))
    return stack[-1]
```

2. When today is warmer than the temperature at the stack top, resolve that older day.
3. The difference in indexes is the wait time.

**Edge cases to check:** empty stack, invalid order, duplicate heights.


```python
# Step-by-step walkthrough:
# Step 1: Keep a stack of indexes whose warmer day has not been found yet.
# Step 2: When today is warmer than the temperature at the stack top, resolve that older
#         day.
# Step 3: The difference in indexes is the wait time.

def daily_temperatures(temperatures):
    # Monotonic decreasing stack of indexes. Time O(n), space O(n).
    res = [0] * len(temperatures)
    stack = []
    for i, temp in enumerate(temperatures):
        while stack and temp > temperatures[stack[-1]]:
            j = stack.pop()
            res[j] = i - j
        stack.append(i)
    return res
```

2. When a shorter bar appears, rectangles using taller bars must end here.
3. A sentinel zero at the end forces all remaining bars to be processed.

**Edge cases to check:** empty stack, invalid order, duplicate heights.


```python
# Step-by-step walkthrough:
# Step 1: Keep an increasing stack of bar heights with their earliest start index.
# Step 2: When a shorter bar appears, rectangles using taller bars must end here.
# Step 3: A sentinel zero at the end forces all remaining bars to be processed.

def largest_rectangle_area(heights):
    # Monotonic stack. Time O(n), space O(n).
    stack = []
    best = 0
    for i, h in enumerate(heights + [0]):
        start = i
        while stack and stack[-1][1] > h:
            index, height = stack.pop()
            best = max(best, height * (i - index))
            start = index
        stack.append((start, h))
    return best
```

**Edge cases to check:** empty input, one element, duplicate values.


```python
# Step-by-step walkthrough:
# Step 1: Find the first index where target could appear using lower bound.
# Step 2: Find the first index greater than target using lower bound on `target + 1`.
# Step 3: The last occurrence is one index before that second boundary.

def search_range(nums, target):
    # Two lower-bound searches. Time O(log n), space O(1).
    def lower_bound(x):
        left, right = 0, len(nums)
        while left < right:
            mid = (left + right) // 2
            if nums[mid] < x:
                left = mid + 1
            else:
                right = mid
        return left

    start = lower_bound(target)
    if start == len(nums) or nums[start] != target:
        return [-1, -1]
    return [start, lower_bound(target + 1) - 1]
```

```python
# Step-by-step walkthrough:
# Step 1: At each step, one half of the array is still sorted.
# Step 2: Check whether the target lies inside the sorted half.
# Step 3: Discard the other half based on that check.

def search_rotated(nums, target):
    # Identify sorted half each step. Time O(log n), space O(1).
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    return -1
```

**How to recognize it:** Look for matrix rows ordered globally.

**How the solution works:**

3. Run normal binary search over the virtual array.

**Edge cases to check:** empty matrix, one row, target absent.


```python
# Step-by-step walkthrough:
# Step 1: Treat the matrix as a flattened sorted array.
# Step 2: Convert a flat index back to row and column with division and modulo.
# Step 3: Run normal binary search over the virtual array.

def search_matrix(matrix, target):
    # Treat matrix as flattened sorted array. Time O(log(mn)), space O(1).
    if not matrix or not matrix[0]:
        return False
    rows, cols = len(matrix), len(matrix[0])
    left, right = 0, rows * cols - 1
    while left <= right:
        mid = (left + right) // 2
        value = matrix[mid // cols][mid % cols]
        if value == target:
            return True
        if value < target:
            left = mid + 1
        else:
            right = mid - 1
    return False
```

```python
# Step-by-step walkthrough:
# Step 1: Move `slow` one step and `fast` two steps.
# Step 2: If there is a cycle, fast eventually catches slow.
# Step 3: If fast reaches the end, there is no cycle.

def has_cycle(head):
    # Floyd fast/slow. Time O(n), space O(1).
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow is fast:
            return True
    return False
```

```python
# Step-by-step walkthrough:
# Step 1: Use a queue for BFS.
# Step 2: For each level, process exactly the number of nodes currently in the queue.
# Step 3: Append children to the queue for the next level.

def level_order(root):
    # BFS by levels. Time O(n), space O(w).
    if not root:
        return []
    q = deque([root])
    res = []
    while q:
        level = []
        for _ in range(len(q)):
            node = q.popleft()
            level.append(node.val)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        res.append(level)
    return res
```

```python
# Step-by-step walkthrough:
# Step 1: Preorder gives the next root.
# Step 2: Inorder tells which values belong to the left and right subtrees.
# Step 3: A hashmap gives each inorder index in O(1).

def build_tree(preorder, inorder):
    # Preorder root + inorder split. Time O(n), space O(n).
    index = {value: i for i, value in enumerate(inorder)}
    pre_i = 0

    def helper(left, right):
        nonlocal pre_i
        if left > right:
            return None
        root_val = preorder[pre_i]
        pre_i += 1
        root = TreeNode(root_val)
        mid = index[root_val]
        root.left = helper(left, mid - 1)
        root.right = helper(mid + 1, right)
        return root

    return helper(0, len(inorder) - 1)
```

**Difficulty:** Hard  
**Pattern:** BFS/DFS  
**Reference:** LeetCode 297  
**Solution to study:** `Codec`

**What the question is testing:** Whether you can recognize `BFS/DFS` and maintain the right invariant while handling edge cases.

**Best approach:** Explore deeply with recursion or a stack, usually marking visited.

**How to recognize it:** Look for paths, components, tree recursion, or exhaustive exploration.

**What to say in an interview:** Start with the brute force idea, then explain why `BFS/DFS` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Serialization stores values in BFS order with `#` for missing children.
2. The null markers preserve the exact tree shape.
3. Deserialization rebuilds children in the same BFS order.

**Edge cases to check:** empty tree/graph, cycles, disconnected parts.


```python
# Step-by-step walkthrough:
# Step 1: Serialization stores values in BFS order with `#` for missing children.
# Step 2: The null markers preserve the exact tree shape.
# Step 3: Deserialization rebuilds children in the same BFS order.

class Codec:
    # BFS serialize/deserialize. Time O(n), space O(n).
    def serialize(self, root):
        if not root:
            return ""
        q = deque([root])
        vals = []
        while q:
            node = q.popleft()
            if node:
                vals.append(str(node.val))
                q.append(node.left)
                q.append(node.right)
            else:
                vals.append("#")
        return ",".join(vals)

    def deserialize(self, data):
        if not data:
            return None
        vals = data.split(",")
        root = TreeNode(int(vals[0]))
        q = deque([root])
        i = 1
        while q:
            node = q.popleft()
            if vals[i] != "#":
                node.left = TreeNode(int(vals[i]))
                q.append(node.left)
            i += 1
            if vals[i] != "#":
                node.right = TreeNode(int(vals[i]))
                q.append(node.right)
            i += 1
        return root
```

**What to say in an interview:** Start with the brute force idea, then explain why `DFS/BFS` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Scan every grid cell.
2. When land is found, count one island and DFS from that cell.
3. The DFS marks connected land as water so it is not counted again.

**Edge cases to check:** empty tree/graph, cycles, disconnected parts.


```python
# Step-by-step walkthrough:
# Step 1: Scan every grid cell.
# Step 2: When land is found, count one island and DFS from that cell.
# Step 3: The DFS marks connected land as water so it is not counted again.

def num_islands(grid):
    # DFS flood fill. Time O(mn), space O(mn) worst case.
    if not grid:
        return 0
    rows, cols = len(grid), len(grid[0])
    count = 0

    def dfs(r, c):
        if r < 0 or c < 0 or r == rows or c == cols or grid[r][c] != "1":
            return
        grid[r][c] = "0"
        dfs(r + 1, c)
        dfs(r - 1, c)
        dfs(r, c + 1)
        dfs(r, c - 1)

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1":
                count += 1
                dfs(r, c)
    return count
```

**What to say in an interview:** Start with the brute force idea, then explain why `Multi-source BFS` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Start from ocean borders instead of every cell.
2. DFS inward only to cells of equal or greater height.
3. Cells reached from both ocean searches are the answer.

**Edge cases to check:** start equals end, empty input, disconnected graph.


```python
# Step-by-step walkthrough:
# Step 1: Start from ocean borders instead of every cell.
# Step 2: DFS inward only to cells of equal or greater height.
# Step 3: Cells reached from both ocean searches are the answer.

def pacific_atlantic(heights):
    # Reverse flow from oceans inward. Time O(mn), space O(mn).
    rows, cols = len(heights), len(heights[0])
    pac, atl = set(), set()

    def dfs(r, c, seen, prev):
        if (r, c) in seen or r < 0 or c < 0 or r == rows or c == cols or heights[r][c] < prev:
            return
        seen.add((r, c))
        for dr, dc in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            dfs(r + dr, c + dc, seen, heights[r][c])

    for c in range(cols):
        dfs(0, c, pac, heights[0][c])
        dfs(rows - 1, c, atl, heights[rows - 1][c])
    for r in range(rows):
        dfs(r, 0, pac, heights[r][0])
        dfs(r, cols - 1, atl, heights[r][cols - 1])
    return list(pac & atl)
```

**How to recognize it:** Look for prerequisites, dependencies, or course ordering.

**What to say in an interview:** Start with the brute force idea, then explain why `Topological Sort` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Build a graph from prerequisite to course.
2. Track how many prerequisites each course still has.
3. Repeatedly take zero-prerequisite courses; if all are taken, no cycle exists.

**Edge cases to check:** cycle, no prerequisites, multiple valid orders.


```python
# Step-by-step walkthrough:
# Step 1: Build a graph from prerequisite to course.
# Step 2: Track how many prerequisites each course still has.
# Step 3: Repeatedly take zero-prerequisite courses; if all are taken, no cycle exists.

def can_finish(num_courses, prerequisites):
    # Kahn topological sort. Time O(V + E), space O(V + E).
    graph = defaultdict(list)
    indegree = [0] * num_courses
    for course, prereq in prerequisites:
        graph[prereq].append(course)
        indegree[course] += 1
    q = deque([i for i, deg in enumerate(indegree) if deg == 0])
    taken = 0
    while q:
        node = q.popleft()
        taken += 1
        for nei in graph[node]:
            indegree[nei] -= 1
            if indegree[nei] == 0:
                q.append(nei)
    return taken == num_courses
```

### Number of Connected Components

**Difficulty:** Medium  
**Pattern:** Union Find/DFS  
**Reference:** LeetCode 323  
**Solution to study:** `count_components`

**What the question is testing:** Whether you can recognize `Union Find/DFS` and maintain the right invariant while handling edge cases.

**Best approach:** Explore deeply with recursion or a stack, usually marking visited.

**How to recognize it:** Look for paths, components, tree recursion, or exhaustive exploration.

**What to say in an interview:** Start with the brute force idea, then explain why `Union Find/DFS` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Start with every node as its own component.
2. Union the two endpoints of each edge.
3. Each successful union reduces the component count by one.

**Edge cases to check:** empty tree/graph, cycles, disconnected parts.


```python
# Step-by-step walkthrough:
# Step 1: Start with every node as its own component.
# Step 2: Union the two endpoints of each edge.
# Step 3: Each successful union reduces the component count by one.

def count_components(n, edges):
    # Union find. Time near O(V + E), space O(V).
    parent = list(range(n))
    rank = [0] * n

    def find(x):
        while x != parent[x]:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x

    def union(a, b):
        ra, rb = find(a), find(b)
        if ra == rb:
            return False
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra] += 1
        return True

    components = n
    for a, b in edges:
        if union(a, b):
            components -= 1
    return components
```

**How the solution works:**

1. Create wildcard buckets like `h*t` to connect one-letter-different words.
2. Run BFS from the begin word.
3. The first time the end word is reached is the shortest transformation length.

**Edge cases to check:** start equals end, empty input, disconnected graph.


```python
# Step-by-step walkthrough:
# Step 1: Create wildcard buckets like `h*t` to connect one-letter-different words.
# Step 2: Run BFS from the begin word.
# Step 3: The first time the end word is reached is the shortest transformation length.

def ladder_length(begin_word, end_word, word_list):
    # BFS over wildcard buckets. Time O(n * m^2), space O(n * m).
    words = set(word_list)
    if end_word not in words:
        return 0
    buckets = defaultdict(list)
    for word in words | {begin_word}:
        for i in range(len(word)):
            buckets[word[:i] + "*" + word[i + 1:]].append(word)
    q = deque([(begin_word, 1)])
    seen = {begin_word}
    while q:
        word, dist = q.popleft()
        if word == end_word:
            return dist
        for i in range(len(word)):
            for nei in buckets[word[:i] + "*" + word[i + 1:]]:
                if nei not in seen:
                    seen.add(nei)
                    q.append((nei, dist + 1))
    return 0
```

1. Keep the lower half in a max heap and upper half in a min heap.
2. Balance the heaps so the lower half has the same count or one extra.
3. The median is either the top of the lower heap or the average of both heap tops.

**Edge cases to check:** ties, k equals 1, k equals n.


```python
# Step-by-step walkthrough:
# Step 1: Keep the lower half in a max heap and upper half in a min heap.
# Step 2: Balance the heaps so the lower half has the same count or one extra.
# Step 3: The median is either the top of the lower heap or the average of both heap
#         tops.

class MedianFinder:
    # Max heap for lower half, min heap for upper half. Add O(log n), find O(1).
    def __init__(self):
        self.low = []
        self.high = []

    def addNum(self, num):
        heapq.heappush(self.low, -num)
        heapq.heappush(self.high, -heapq.heappop(self.low))
        if len(self.high) > len(self.low):
            heapq.heappush(self.low, -heapq.heappop(self.high))

    def findMedian(self):
        if len(self.low) > len(self.high):
            return -self.low[0]
        return (-self.low[0] + self.high[0]) / 2
```

### Task Scheduler

**Difficulty:** Medium  
**Pattern:** Heap + Greedy  
**Reference:** LeetCode 621  
**Solution to study:** `least_interval`

**What the question is testing:** Whether you can recognize `Heap + Greedy` and maintain the right invariant while handling edge cases.

**Best approach:** Use a priority queue when repeatedly asking for top, kth, smallest, or largest.

**How to recognize it:** Look for top k, kth largest, closest, or merge k sorted streams.

**What to say in an interview:** Start with the brute force idea, then explain why `Heap + Greedy` removes wasted work. State the invariant before coding.

**How the solution works:**

1. The most frequent task determines the minimum frame length.
2. Arrange the most frequent task with cooldown gaps.
3. The answer is the larger of actual task count and required frame size.

**Edge cases to check:** ties, k equals 1, k equals n.


```python
# Step-by-step walkthrough:
# Step 1: The most frequent task determines the minimum frame length.
# Step 2: Arrange the most frequent task with cooldown gaps.
# Step 3: The answer is the larger of actual task count and required frame size.

def least_interval(tasks, n):
    # Greedy formula. Time O(t), space O(1).
    counts = Counter(tasks).values()
    max_count = max(counts)
    num_max = sum(1 for c in counts if c == max_count)
    return max(len(tasks), (max_count - 1) * (n + 1) + num_max)
```

### Implement Trie

**Difficulty:** Medium  
**Pattern:** Core  
**Reference:** LeetCode 208  
**Solution to study:** `Trie`

**What the question is testing:** Whether you can recognize `Core` and maintain the right invariant while handling edge cases.

**Best approach:** Use the named pattern from the study plan and keep the invariant explicit.

**How to recognize it:** Match the problem wording to the pattern column.

**What to say in an interview:** Start with the brute force idea, then explain why `Core` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Each node stores child pointers by character.
2. Insert walks or creates nodes for each character.
3. Search and prefix lookup walk the same path and check either `is_end` or existence.

**Edge cases to check:** empty input, one element, duplicate values.


```python
# Step-by-step walkthrough:
# Step 1: Each node stores child pointers by character.
# Step 2: Insert walks or creates nodes for each character.
# Step 3: Search and prefix lookup walk the same path and check either `is_end` or
#         existence.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

class Trie:
    # Insert/search/prefix are O(word length).
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True

    def search(self, word):
        node = self._find(word)
        return bool(node and node.is_end)

    def startsWith(self, prefix):
        return self._find(prefix) is not None

    def _find(self, text):
        node = self.root
        for ch in text:
            if ch not in node.children:
                return None
            node = node.children[ch]
        return node
```

### Design Add and Search Words

**Pattern:** Trie + DFS  
**Reference:** LeetCode 211  
**Solution to study:** `WordDictionary`

**What the question is testing:** Whether you can recognize `Trie + DFS` and maintain the right invariant while handling edge cases.

**Best approach:** Explore deeply with recursion or a stack, usually marking visited.

**How to recognize it:** Look for paths, components, tree recursion, or exhaustive exploration.

**What to say in an interview:** Start with the brute force idea, then explain why `Trie + DFS` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Store words in a trie.
2. Normal letters follow one child.
3. A dot branches DFS into every child at that position.

**Edge cases to check:** empty tree/graph, cycles, disconnected parts.


```python
# Step-by-step walkthrough:
# Step 1: Store words in a trie.
# Step 2: Normal letters follow one child.
# Step 3: A dot branches DFS into every child at that position.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

class WordDictionary:
    # Trie + DFS for '.' wildcard.
    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word):
        node = self.root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True

    def search(self, word):
        def dfs(i, node):
            if i == len(word):
                return node.is_end
            ch = word[i]
            if ch == ".":
                return any(dfs(i + 1, child) for child in node.children.values())
            return ch in node.children and dfs(i + 1, node.children[ch])
        return dfs(0, self.root)
```

### Word Search II

**Difficulty:** Hard  
**Pattern:** Trie + Backtrack  
**Reference:** LeetCode 212  
**Solution to study:** `find_words`

**What the question is testing:** Whether you can recognize `Trie + Backtrack` and maintain the right invariant while handling edge cases.

**Best approach:** Store words as character paths.

**How to recognize it:** Look for prefix lookup, autocomplete, wildcard word search, or many dictionary words.

**What to say in an interview:** Start with the brute force idea, then explain why `Trie + Backtrack` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Build a trie from all words.
2. Backtrack from each board cell while following trie edges.
3. When a trie node has a full word, add it and clear it to avoid duplicates.

**Edge cases to check:** prefix that is not a word, duplicate words, dot wildcard.


```python
# Step-by-step walkthrough:
# Step 1: Build a trie from all words.
# Step 2: Backtrack from each board cell while following trie edges.
# Step 3: When a trie node has a full word, add it and clear it to avoid duplicates.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

def find_words(board, words):
    # Trie prunes backtracking. Time bounded by board paths, space O(total word chars).
    root = TrieNode()
    for word in words:
        node = root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.word = word

    rows, cols = len(board), len(board[0])
    res = []

    def dfs(r, c, node):
        if r < 0 or c < 0 or r == rows or c == cols:
            return
        ch = board[r][c]
        if ch == "#" or ch not in node.children:
            return
        nxt = node.children[ch]
        if nxt.word:
            res.append(nxt.word)
            nxt.word = None
        board[r][c] = "#"
        dfs(r + 1, c, nxt)
        dfs(r - 1, c, nxt)
        dfs(r, c + 1, nxt)
        dfs(r, c - 1, nxt)
        board[r][c] = ch

    for r in range(rows):
        for c in range(cols):
            dfs(r, c, root)
    return res
```

**Difficulty:** Easy  
**Pattern:** 1D DP  
**Reference:** LeetCode 70  
**Solution to study:** `climb_stairs`

**How to recognize it:** Look for choose/skip, count ways, min coins, or possible segmentation.

**What to say in an interview:** Start with the brute force idea, then explain why `1D DP` removes wasted work. State the invariant before coding.

**How the solution works:**

1. This is Fibonacci in disguise.
2. Ways to reach the current step equals previous step plus two steps back.
3. Only two previous values are needed.

**Edge cases to check:** zero amount, empty prefix, impossible state.


```python
# Step-by-step walkthrough:
# Step 1: This is Fibonacci in disguise.
# Step 2: Ways to reach the current step equals previous step plus two steps back.
# Step 3: Only two previous values are needed.

def climb_stairs(n):
    # Fibonacci-style DP. Time O(n), space O(1).
    if n <= 2:
        return n
    a, b = 1, 2
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b
```

2. Robbing it means adding its value to the best answer two houses back.

### Coin Change

**What to say in an interview:** Start with the brute force idea, then explain why `1D DP` removes wasted work. State the invariant before coding.

**How the solution works:**

1. `dp[x]` stores the fewest coins needed for amount x.
2. For each amount, try every coin as the last coin used.
3. Unreachable amounts stay infinity and become `-1` at the end.

**Edge cases to check:** zero amount, empty prefix, impossible state.


```python
# Step-by-step walkthrough:
# Step 1: `dp[x]` stores the fewest coins needed for amount x.
# Step 2: For each amount, try every coin as the last coin used.
# Step 3: Unreachable amounts stay infinity and become `-1` at the end.

def coin_change(coins, amount):
    # Unbounded knapsack min coins. Time O(amount * coins), space O(amount).
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0
    for total in range(1, amount + 1):
        for coin in coins:
            if total >= coin:
                dp[total] = min(dp[total], dp[total - coin] + 1)
    return dp[amount] if dp[amount] != float("inf") else -1
```

### Longest Increasing Subsequence

**What to say in an interview:** Start with the brute force idea, then explain why `1D DP` removes wasted work. State the invariant before coding.

**How the solution works:**

1. `tails[i]` stores the smallest possible tail for an increasing subsequence of length i+1.
2. Binary search finds where the current number should replace or extend a tail.
3. The length of `tails` is the LIS length.

**Edge cases to check:** zero amount, empty prefix, impossible state.


```python
# Step-by-step walkthrough:
# Step 1: `tails[i]` stores the smallest possible tail for an increasing subsequence of
#         length i+1.
# Step 2: Binary search finds where the current number should replace or extend a tail.
# Step 3: The length of `tails` is the LIS length.

def length_of_lis(nums):
    # Patience sorting tails. Time O(n log n), space O(n).
    tails = []
    for x in nums:
        i = bisect_left(tails, x)
        if i == len(tails):
            tails.append(x)
        else:
            tails[i] = x
    return len(tails)
```

**How the solution works:**

1. Each cell can be reached from the top or from the left.
2. The 1D array stores the number of ways to reach each column in the current row.
3. Updating left to right adds the left value into the current column.

```python
# Step-by-step walkthrough:
# Step 1: Each cell can be reached from the top or from the left.
# Step 2: The 1D array stores the number of ways to reach each column in the current
#         row.
# Step 3: Updating left to right adds the left value into the current column.

def unique_paths(m, n):
    # 2D paths optimized to 1D. Time O(mn), space O(n).
    dp = [1] * n
    for _ in range(1, m):
        for c in range(1, n):
            dp[c] += dp[c - 1]
    return dp[-1]
```

**What to say in an interview:** Start with the brute force idea, then explain why `2D DP` removes wasted work. State the invariant before coding.

**How the solution works:**

2. If they match, extend the diagonal previous result.
3. If not, keep the best result from skipping one character from either string.

**Edge cases to check:** empty row/string, first row/column.


```python
# Step-by-step walkthrough:
# Step 1: Compare characters from both strings.
# Step 2: If they match, extend the diagonal previous result.
# Step 3: If not, keep the best result from skipping one character from either string.

def longest_common_subsequence(text1, text2):
    # 2D DP optimized to 1D. Time O(mn), space O(n).
    dp = [0] * (len(text2) + 1)
    for a in text1:
        prev = 0
        for j, b in enumerate(text2, 1):
            temp = dp[j]
            if a == b:
                dp[j] = prev + 1
            else:
                dp[j] = max(dp[j], dp[j - 1])
            prev = temp
    return dp[-1]
```

2. Try every split point `j` before `i`.
3. If `dp[j]` is true and `s[j:i]` is a word, then `dp[i]` is true.

### Edit Distance

**What to say in an interview:** Start with the brute force idea, then explain why `2D DP` removes wasted work. State the invariant before coding.

**How the solution works:**

1. The DP compares prefixes of both words.
2. Matching characters cost nothing beyond the diagonal state.
3. Different characters take one operation plus the best insert, delete, or replace option.

**Edge cases to check:** empty row/string, first row/column.


```python
# Step-by-step walkthrough:
# Step 1: The DP compares prefixes of both words.
# Step 2: Matching characters cost nothing beyond the diagonal state.
# Step 3: Different characters take one operation plus the best insert, delete, or
#         replace option.

def min_distance(word1, word2):
    # Edit distance. Time O(mn), space O(n).
    prev = list(range(len(word2) + 1))
    for i, a in enumerate(word1, 1):
        curr = [i] + [0] * len(word2)
        for j, b in enumerate(word2, 1):
            if a == b:
                curr[j] = prev[j - 1]
            else:
                curr[j] = 1 + min(prev[j], curr[j - 1], prev[j - 1])
        prev = curr
    return prev[-1]
```

**What to say in an interview:** Start with the brute force idea, then explain why `Sorting` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Split the array until each part has at most one item.
2. Recursively sort both halves.
3. Merge the two sorted halves by repeatedly taking the smaller front item.

**Edge cases to check:** duplicates, already sorted, reverse sorted.


```python
# Step-by-step walkthrough:
# Step 1: Split the array until each part has at most one item.
# Step 2: Recursively sort both halves.
# Step 3: Merge the two sorted halves by repeatedly taking the smaller front item.

def merge_sort(arr):
    # Stable divide and conquer. Time O(n log n), space O(n).
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    return res + left[i:] + right[j:]
```

### Implement quicksort

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** Practice item  
**Solution to study:** `quick_sort`

**What the question is testing:** Whether you can recognize `Sorting` and maintain the right invariant while handling edge cases.

**Best approach:** Sort first to create order, then scan, merge, or use pointers.

**How to recognize it:** Look for intervals, groups, nearest values, or custom ordering.

**What to say in an interview:** Start with the brute force idea, then explain why `Sorting` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Choose a pivot and partition the array around it.
2. Values less than or equal to pivot move before it.
3. Recursively sort the left and right partitions.

**Edge cases to check:** duplicates, already sorted, reverse sorted.


```python
# Step-by-step walkthrough:
# Step 1: Choose a pivot and partition the array around it.
# Step 2: Values less than or equal to pivot move before it.
# Step 3: Recursively sort the left and right partitions.

def quick_sort(arr):
    # In-place quicksort. Average O(n log n), worst O(n^2), space O(log n).
    def partition(low, high):
        pivot = arr[high]
        i = low
        for j in range(low, high):
            if arr[j] <= pivot:
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
        arr[i], arr[high] = arr[high], arr[i]
        return i

    def sort(low, high):
        if low < high:
            p = partition(low, high)
            sort(low, p - 1)
            sort(p + 1, high)

    sort(0, len(arr) - 1)
    return arr
```

### Sort Colors

**What the question is testing:** Whether you can recognize `Sorting` and maintain the right invariant while handling edge cases.

**How the solution works:**

1. Maintain three regions: zeros on the left, twos on the right, unknown in the middle.
2. `mid` scans the unknown region.
3. Swap zeros left and twos right until all values are placed.

**Edge cases to check:** duplicates, already sorted, reverse sorted.


```python
# Step-by-step walkthrough:
# Step 1: Maintain three regions: zeros on the left, twos on the right, unknown in the
#         middle.
# Step 2: `mid` scans the unknown region.
# Step 3: Swap zeros left and twos right until all values are placed.

def sort_colors(nums):
    # Dutch National Flag. Time O(n), space O(1).
    low = mid = 0
    high = len(nums) - 1
    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
```

### Merge Intervals

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** LeetCode 56  
**Solution to study:** `merge_intervals`

**What the question is testing:** Whether you can recognize `Sorting` and maintain the right invariant while handling edge cases.

**Best approach:** Sort first to create order, then scan, merge, or use pointers.

**How to recognize it:** Look for intervals, groups, nearest values, or custom ordering.

**What to say in an interview:** Start with the brute force idea, then explain why `Sorting` removes wasted work. State the invariant before coding.

**How the solution works:**

1. Sort intervals by start time.
2. If the next interval starts after the previous end, start a new interval.
3. Otherwise, extend the previous end if needed.

**Edge cases to check:** duplicates, already sorted, reverse sorted.


```python
# Step-by-step walkthrough:
# Step 1: Sort intervals by start time.
# Step 2: If the next interval starts after the previous end, start a new interval.
# Step 3: Otherwise, extend the previous end if needed.

def merge_intervals(intervals):
    # Sort by start, merge overlaps. Time O(n log n), space O(n).
    intervals.sort()
    merged = []
    for start, end in intervals:
        if not merged or start > merged[-1][1]:
            merged.append([start, end])
        else:
            merged[-1][1] = max(merged[-1][1], end)
    return merged
```
