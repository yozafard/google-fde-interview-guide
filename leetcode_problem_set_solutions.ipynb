### Two Sum

**Difficulty:** Easy  
**Pattern:** Hash Map  
**Reference:** LeetCode 1  
**Solution to study:** `two_sum`

**Question in plain English:** Given a list of numbers and a target, return the two indexes whose numbers add up to the target.

**Like you are 12:** Pretend each number is a puzzle piece. For every piece, you ask, "What other piece would complete the target?" A dictionary is your memory box for pieces you already saw.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Map` idea and explain what each variable is remembering.

**Best approach:** For `Two Sum`, use the `Hash Map` pattern. Use a hash map when each item needs to remember or find related information quickly. **How the solution works:** 

1. Name the state first. In `two_sum`, the state is the small amount of memory that makes `Hash Map` work for `Two Sum`.
2. Create a `seen` dictionary where each number points to its index. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.
3. For every number, compute the complement `target - x`. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.
4. If the complement is already in `seen`, the pair is found; otherwise store the current number. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.

**Beginner checks:** duplicates, empty input, repeated values.


```python
# Defines `two_sum`. The values inside the parentheses are the inputs: `nums, target`.
def two_sum(nums, target):
    # Hash map. Time O(n), space O(n).
    # Creates an empty dictionary. In this solution it is the memory of values already processed.
    seen = {}
    # Loops through the input while giving both the index and the value at that index.
    for i, x in enumerate(nums):
        # Computes the complement: the exact number that would pair with `x` to hit `target`.
        need = target - x
        # Asks whether the complement was already seen earlier. If yes, the pair is complete.
        if need in seen:
            # Returns the answer produced by the previous lines.
            return [seen[need], i]
        # Stores the current number and its index so a later number can find it quickly.
        seen[x] = i
    # Returns an empty answer when no valid result was found.
    return []
```

In `merge_intervals`, I will update that state as I scan/process the input, return when the success condition is reached, and then test a normal case plus edge cases like empty input, one item, duplicates, or boundary values when relevant."

**How the solution works:** 

1. Name the state first. In `merge_intervals`, the state is the small amount of memory that makes `Sorting` work for `Merge Intervals`.
2. Sort intervals by start time. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.
3. If the next interval starts after the previous end, start a new interval. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.
4. Otherwise, extend the previous end if needed. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.
5. Once all required input has been processed, return the answer. At that point, the algorithm has considered every case the pattern says it needs to consider.

**Beginner checks:** duplicates, already sorted, reverse sorted.


```python
# Defines `merge_intervals`. The values inside the parentheses are the inputs: `intervals`.
def merge_intervals(intervals):
    # Sort by start, merge overlaps. Time O(n log n), space O(n).
    # Calls a helper or method that performs one action.
    intervals.sort()
    # Stores a value in `merged` so later lines can use it.
    merged = []
    # Visits each relevant item one at a time.
    for start, end in intervals:
        # Checks whether this specific case is happening before choosing the next action.
        if not merged or start > merged[-1][1]:
            # Adds a new item to `merged`. For lists/stacks/queues, this records work or part of the
            # answer.
            merged.append([start, end])
        # Handles the remaining case after the earlier checks failed.
        else:
            # Stores a value in `merged[-1][1]` so later lines can use it.
            merged[-1][1] = max(merged[-1][1], end)
    # Returns the answer produced by the previous lines.
    return merged
```
