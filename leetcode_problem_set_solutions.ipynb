````python
# Shared imports and node classes used by many solutions below.
from collections import Counter, defaultdict, deque  # Counter=tally; defaultdict=auto-default dict; deque=O(1) ends
from bisect import bisect_left                        # binary-search insertion point into a sorted list
import heapq                                          # binary heap (priority queue) helpers


# Singly linked list node: a value plus a pointer to the next node.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val      # the data stored in this node
        self.next = next    # reference to the following node (None at the tail)


# Binary tree node: a value plus pointers to left and right children.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val      # the data stored in this node
        self.left = left     # left child (None if absent)
        self.right = right   # right child (None if absent)


# Graph node: a value plus a list of adjacent nodes (used by Clone Graph).
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        # `neighbors or []` would break on a real empty list, so test for None explicitly
        self.neighbors = neighbors if neighbors is not None else []
````

```
    for i, x in nums:
```

````python
def two_sum(nums, target):
    # seen = value -> index of every number we've already passed
    seen = {}
    for i, x in enumerate(nums):
        need = target - x          # the partner value that would complete the pair
        if need in seen:           # have we already seen that partner?
            return [seen[need], i]  # yes -> its earlier index plus the current one
        seen[x] = i                # otherwise remember this value for a future partner
    return []                      # no pair adds up to target
````

````python
def group_anagrams(strs):
    # groups: sorted-letters signature -> list of words sharing it
    groups = defaultdict(list)
    for word in strs:
        # anagrams sort to the same string, so use that as the bucket key
        groups["".join(sorted(word))].append(word)
    return list(groups.values())   # we only want the buckets, not the keys
````

```
    # pick the k values with the largest counts

```

**Beginner checks:** duplicates, empty input, repeated values.


````python
def top_k_frequent(nums, k):
    count = Counter(nums)   # value -> how many times it appears
    # nlargest picks the k keys with the highest count (key=count.get reads each key's count)
    return heapq.nlargest(k, count.keys(), key=count.get)
````

### Product of Array Except Self

**Difficulty:** Medium  
**Pattern:** Prefix/Suffix  
**Reference:** LeetCode 238  
**Solution to study:** `product_except_self`

**Question in plain English:** Given numbers, return a new list where each position has the product of every other number except itself.

**Like you are 12:** For each spot, multiply everything on its left and everything on its right. Do not use the number sitting in that spot.

**What the question is testing:** Whether you can turn the problem into a simple `Prefix/Suffix` idea and explain what each variable is remembering.

**Best approach:** For `Product of Array Except Self`, use the `Prefix/Suffix` pattern. Use prefix and suffix products when each position needs information from both sides but cannot use itself. This avoids recomputing products for every index. Start by describing the brute force idea, then explain what work this approach avoids. That gives the interviewer a clear before-and-after story.

**What to say in an interview:** I would say: "Let me first restate `Product of Array Except Self` in simple terms and mention the brute force approach. In `product_except_self`, I will update that state as I scan/process the input, return when the success condition is reached, and then test a normal case plus edge cases like empty input, one item, duplicates, or boundary values when relevant."

**How the solution works:**

```
function product_except_self(nums):
    res = array of 1s
    prefix = 1
    for i in 0..n-1:                 # left pass
        res[i] = prefix              # product of everything left of i
        prefix *= nums[i]
    suffix = 1
    for i in n-1..0:                 # right pass
        res[i] *= suffix             # multiply in product of everything right of i
        suffix *= nums[i]
    return res
```

**Beginner checks:** empty input, one element, duplicate values.


````python
def product_except_self(nums):
    # res[i] will become the product of everything except nums[i], built without division
    res = [1] * len(nums)
    prefix = 1                     # running product of all elements to the LEFT of i
    for i, x in enumerate(nums):
        res[i] = prefix            # left part only (suffix multiplied in later)
        prefix *= x                # extend the left product to include x
    suffix = 1                     # running product of all elements to the RIGHT of i
    for i in range(len(nums) - 1, -1, -1):   # sweep right-to-left
        res[i] *= suffix           # combine the stored left product with the right product
        suffix *= nums[i]          # extend the right product to include nums[i]
    return res
````

```
function is_palindrome(s):
    left, right = 0, n-1
    while left < right:
        skip non-alphanumeric chars from the left
        skip non-alphanumeric chars from the right
        if lower(s[left]) != lower(s[right]):
            return False
        left += 1; right -= 1
    return True
```

**Beginner checks:** duplicates, empty input, pointers crossing.


````python
def is_palindrome(s):
    left, right = 0, len(s) - 1     # two pointers closing in from both ends
    while left < right:
        # skip anything that is not a letter/digit on the left side
        while left < right and not s[left].isalnum():
            left += 1
        # ...and on the right side
        while left < right and not s[right].isalnum():
            right -= 1
        # compare the two meaningful characters case-insensitively
        if s[left].lower() != s[right].lower():
            return False           # mismatch -> not a palindrome
        left += 1                  # both matched, step inward
        right -= 1
    return True                    # pointers met without a mismatch
````

### 3Sum

**How the solution works:**

```
function three_sum(nums):
    sort(nums)
    res = []
    for i in 0..n-3:
        if nums[i] == nums[i-1]: continue        # skip duplicate first value
        left, right = i+1, n-1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                record triple
                move left/right inward, skipping duplicate values
            else if total < 0: left += 1         # raise the low number
            else:              right -= 1         # lower the high number
    return res
```

**Beginner checks:** duplicates, empty input, pointers crossing.


````python
def three_sum(nums):
    nums.sort()                    # sorting lets us use two pointers and skip duplicates
    res = []
    for i in range(len(nums) - 2):     # i fixes the first number of the triple
        # skip a repeated first value so we don't emit duplicate triples
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        left, right = i + 1, len(nums) - 1   # two-pointer search in the rest
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
                # skip duplicate seconds/thirds so each triple is unique
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            elif total < 0:
                left += 1          # sum too small -> raise the lower number
            else:
                right -= 1         # sum too big -> lower the higher number
    return res
````

```
    best = 0
    while left < right:
        best = max(best, min(height[left], height[right]) * (right-left))
```

````python
def max_area(height):
    left, right = 0, len(height) - 1   # widest possible container to start
    best = 0
    while left < right:
        # area = shorter wall * distance between the walls
        best = max(best, min(height[left], height[right]) * (right - left))
        # moving the taller wall can't help (width shrinks, height capped by shorter);
        # move the shorter wall inward hoping for a taller one
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    return best
````

```
function trap(height):
    left, right = 0, n-1
    left_max = right_max = water = 0
    while left < right:
        if height[left] < height[right]:        # lower side is the binding wall
            left_max = max(left_max, height[left])
            water += left_max - height[left]     # water above this column

```

````python
def trap(height):
    left, right = 0, len(height) - 1
    left_max = right_max = water = 0   # tallest wall seen from each side; total trapped water
    while left < right:
        # work on whichever side has the lower wall: that side's max is the true bound
        if height[left] < height[right]:
            left_max = max(left_max, height[left])   # tallest wall to the left so far
            water += left_max - height[left]         # water above this column = bound - its height
            left += 1
        else:
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1
    return water
````

````python
class Codec:
    # Serialize via BFS, marking missing children with "#".
    def serialize(self, root):
        if not root:
            return ""
        q = deque([root])
        vals = []
        while q:
            node = q.popleft()
            if node:
                vals.append(str(node.val))
                q.append(node.left)    # enqueue children (possibly None)
                q.append(node.right)
            else:
                vals.append("#")       # placeholder for a missing child
        return ",".join(vals)
    # Rebuild by reading the same BFS order back.
    def deserialize(self, data):
        if not data:
            return None
        vals = data.split(",")
        root = TreeNode(int(vals[0]))
        q = deque([root])
        i = 1                          # index of the next child value to attach
        while q:
            node = q.popleft()
            if vals[i] != "#":         # left child present
                node.left = TreeNode(int(vals[i]))
                q.append(node.left)
            i += 1
            if vals[i] != "#":         # right child present
                node.right = TreeNode(int(vals[i]))
                q.append(node.right)
            i += 1
        return root
````

````python
def pacific_atlantic(heights):
    rows, cols = len(heights), len(heights[0])
    pac, atl = set(), set()        # cells from which water can reach each ocean
    # walk UPHILL from an ocean edge: water flows down, so reachability reverses
    def dfs(r, c, seen, prev):
        # stop if revisited, off-grid, or the next cell is lower than where we came from
        if (r, c) in seen or r < 0 or c < 0 or r == rows or c == cols or heights[r][c] < prev:
            return
        seen.add((r, c))
        for dr, dc in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            dfs(r + dr, c + dc, seen, heights[r][c])
    # start from every border cell touching each ocean
    for c in range(cols):
        dfs(0, c, pac, heights[0][c])              # top row -> Pacific
        dfs(rows - 1, c, atl, heights[rows - 1][c])# bottom row -> Atlantic
    for r in range(rows):
        dfs(r, 0, pac, heights[r][0])              # left col -> Pacific
        dfs(r, cols - 1, atl, heights[r][cols - 1])# right col -> Atlantic
    return list(pac & atl)         # cells that can drain to BOTH oceans
````

```
    build graph: prereq -> dependent courses
    indegree[c] = number of prereqs of c
    q = all courses with indegree 0
    taken = 0
```

````python
def can_finish(num_courses, prerequisites):
    # Kahn's topological sort: a course is takeable once all its prereqs are done.
    graph = defaultdict(list)      # prereq -> courses that depend on it
    indegree = [0] * num_courses   # how many prereqs each course still needs
    for course, prereq in prerequisites:
        graph[prereq].append(course)
        indegree[course] += 1
    q = deque([i for i, deg in enumerate(indegree) if deg == 0])   # courses with no prereqs
    taken = 0
    while q:
        node = q.popleft()
        taken += 1                 # take this course
        for nei in graph[node]:    # one prereq satisfied for each dependent course
            indegree[nei] -= 1
            if indegree[nei] == 0: # all prereqs met -> it becomes takeable
                q.append(nei)
    return taken == num_courses    # all taken means no cycle blocked us
````

**How the solution works:**

```
function count_components(n, edges):           # Union-Find
    parent[x] = x; rank[x] = 0
    find(x):  follow parent to the root, compressing the path
    union(a,b): link roots by rank; return False if already joined
    components = n
    for (a,b) in edges:
        if union(a,b): components -= 1          # each merge joins two groups
    return components
```

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


````python
def count_components(n, edges):
    # Union-Find: merge connected nodes, count how many separate groups remain.
    parent = list(range(n))        # parent[x] = x's representative (starts as itself)
    rank = [0] * n                 # approximate tree height, to keep unions shallow
    def find(x):                   # find x's root, with path compression
        while x != parent[x]:
            parent[x] = parent[parent[x]]   # point x at its grandparent (flattens the tree)
            x = parent[x]
        return x
    def union(a, b):               # merge the groups of a and b; return False if already joined
        ra, rb = find(a), find(b)
        if ra == rb:
            return False
        # attach the shorter tree under the taller one
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra] += 1
        return True
    components = n                 # start with every node in its own group
    for a, b in edges:
        if union(a, b):            # a successful merge joins two groups into one
            components -= 1
    return components
````

````python
def ladder_length(begin_word, end_word, word_list):
    words = set(word_list)         # fast membership tests
    if end_word not in words:
        return 0                   # impossible if the target isn't a valid word
    # buckets: a pattern like "h*t" -> all words that match it (differ by one letter)
    buckets = defaultdict(list)
    for word in words | {begin_word}:
        for i in range(len(word)):
            buckets[word[:i] + "*" + word[i + 1:]].append(word)
    q = deque([(begin_word, 1)])   # BFS over words; distance = transformation count
    seen = {begin_word}
    while q:
        word, dist = q.popleft()
        if word == end_word:
            return dist            # BFS guarantees this is the shortest ladder
        for i in range(len(word)): # neighbors share a wildcard pattern with word
            for nei in buckets[word[:i] + "*" + word[i + 1:]]:
                if nei not in seen:
                    seen.add(nei)
                    q.append((nei, dist + 1))
    return 0                       # target unreachable
````

```
        if low bigger: return low's top            # odd count
        else: return average of low.top and high.top   # even count
```

**Beginner checks:** ties, k equals 1, k equals n.


````python
class MedianFinder:
    # Two heaps: `low` (max-heap via negatives) holds the smaller half,
    # `high` (min-heap) holds the larger half. The median sits at their tops.
    def __init__(self):
        self.low = []              # max-heap (store negatives) of the smaller half
        self.high = []             # min-heap of the larger half
    def addNum(self, num):
        heapq.heappush(self.low, -num)               # tentatively add to the low half
        heapq.heappush(self.high, -heapq.heappop(self.low))   # move low's max over to high (keeps order)
        if len(self.high) > len(self.low):           # rebalance so low has >= high
            heapq.heappush(self.low, -heapq.heappop(self.high))
    def findMedian(self):
        if len(self.low) > len(self.high):           # odd count -> low holds the middle
            return -self.low[0]
        return (-self.low[0] + self.high[0]) / 2     # even count -> average the two middles
````

**How the solution works:**

```
class Trie:
    node = { children: map char->node, is_end: bool }
    insert(word):
        walk/create a path of nodes for each char; mark last node is_end = True
    search(word):
        follow the path; True only if it exists AND last node is_end
    startsWith(prefix):
        follow the path; True if it exists (is_end not required)
```

**Beginner checks:** empty input, one element, duplicate values.


````python
class TrieNode:
    def __init__(self):
        self.children = {}         # next character -> child TrieNode
        self.word = None           # (used by Word Search II) full word ending here
        self.is_end = False        # True if a word ends exactly at this node


class Trie:
    def __init__(self):
        self.root = TrieNode()     # empty root represents the empty prefix
    def insert(self, word):
        node = self.root
        for ch in word:
            # walk down, creating missing child nodes as needed
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True         # mark the final node as a complete word
    def search(self, word):
        node = self._find(word)    # full word counts only if its path exists AND ends a word
        return bool(node and node.is_end)
    def startsWith(self, prefix):
        return self._find(prefix) is not None   # prefix just needs the path to exist
    def _find(self, text):         # walk the trie; return the end node or None
        node = self.root
        for ch in text:
            if ch not in node.children:
                return None        # path breaks -> text isn't present
            node = node.children[ch]
        return node
````

**How the solution works:**

```
function find_words(board, words):                # Trie + DFS backtracking
    build a trie of all words; store the full word at its end node
    res = []
    define dfs(r, c, node):
        if off board, or cell used ("#"), or cell char not a child: return
        nxt = node.children[char]
        if nxt holds a complete word: add it to res; clear it (avoid duplicates)
        mark cell "#"; dfs into 4 neighbors with nxt; restore cell
    run dfs(r, c, root) from every cell
    return res
```

**Beginner checks:** prefix that is not a word, duplicate words, dot wildcard.


````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None           # holds the full word at the node where it ends
        self.is_end = False


def find_words(board, words):
    # Build a trie of all target words so one board walk can match them all at once.
    root = TrieNode()
    for word in words:
        node = root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.word = word           # stash the whole word at its terminal node
    rows, cols = len(board), len(board[0])
    res = []
    def dfs(r, c, node):
        if r < 0 or c < 0 or r == rows or c == cols:
            return                 # off the board
        ch = board[r][c]
        if ch == "#" or ch not in node.children:
            return                 # already used this cell, or no word continues with ch
        nxt = node.children[ch]    # advance down the trie
        if nxt.word:               # a complete word ends here
            res.append(nxt.word)
            nxt.word = None        # clear it so we don't report the same word twice
        board[r][c] = "#"          # mark cell used for this path
        dfs(r + 1, c, nxt)         # explore 4 neighbors, continuing down the trie
        dfs(r - 1, c, nxt)
        dfs(r, c + 1, nxt)
        dfs(r, c - 1, nxt)
        board[r][c] = ch           # restore the cell when backtracking
    for r in range(rows):
        for c in range(cols):
            dfs(r, c, root)        # try starting a word at every cell
    return res
````

**How the solution works:**

```
function merge_sort(arr):
    if len(arr) <= 1: return arr
    split into left and right halves; merge_sort each
    merge: repeatedly take the smaller front element of the two sorted halves
    append any remaining elements
    return merged list
```

**Beginner checks:** duplicates, already sorted, reverse sorted.


````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr                 # 0 or 1 element is already sorted
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])   # sort each half recursively
    right = merge_sort(arr[mid:])
    # merge the two sorted halves into one sorted list
    res = []
    i = j = 0                      # read pointers into left and right
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:    # take the smaller front element (<= keeps it stable)
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    return res + left[i:] + right[j:]   # append whatever remains in either half
````

**How the solution works:**

```

    define partition(low, high):           # Lomuto
        pivot = arr[high]; i = low
        for j in low..high-1:
            if arr[j] <= pivot: swap arr[i], arr[j]; i += 1
        swap arr[i], arr[high]; return i    # pivot's final spot
    define sort(low, high):
        if low < high:
            p = partition(low, high)
            sort(low, p-1); sort(p+1, high)  # recurse on both sides

```

**Beginner checks:** duplicates, already sorted, reverse sorted.


````python
def quick_sort(arr):
    # Lomuto partition: place pivot, put smaller elements to its left.
    def partition(low, high):
        pivot = arr[high]          # choose the last element as the pivot
        i = low                    # boundary: arr[low..i-1] are all <= pivot
        for j in range(low, high):
            if arr[j] <= pivot:
                arr[i], arr[j] = arr[j], arr[i]   # move a small element into the left zone
                i += 1
        arr[i], arr[high] = arr[high], arr[i]     # drop pivot into its final sorted spot
        return i                   # pivot's final index
    def sort(low, high):
        if low < high:             # more than one element in this range
            p = partition(low, high)
            sort(low, p - 1)       # sort the side smaller than the pivot
            sort(p + 1, high)      # sort the side larger than the pivot
    sort(0, len(arr) - 1)
    return arr
````
