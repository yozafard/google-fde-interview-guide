### Two Sum

**Difficulty:** Easy

**Question in plain English:** Given a list of numbers and a target, return the two indexes whose numbers add up to the target.

**Like you are 12:** Pretend each number is a puzzle piece. For every piece, you ask, "What other piece would complete the target?" A dictionary is your memory box for pieces you already saw.

3. For every number, compute the complement `target - x`. Think of this as one mechanical action: read the current item, compare it with the saved state, then update the state if needed.

```python
# Beginner walkthrough:
# Read this section slowly. Each note explains the next meaningful code line, in the same order the code appears.
# Line 1: `def two_sum(nums, target):` defines `two_sum`. The values inside the parentheses are the inputs:
#        `nums, target`.
# Line 2: `seen = {}` creates an empty dictionary. In this solution it is the memory of values already
#        processed.
# Line 3: `for i, x in enumerate(nums):` loops through the input while giving both the index and the value at
#        that index.
# Line 4: `need = target - x` computes the complement: the exact number that would pair with `x` to hit
#        `target`.
# Line 5: `if need in seen:` asks whether the complement was already seen earlier. If yes, the pair is complete.
# Line 6: `return [seen[need], i]` returns the answer produced by the previous lines.
# Line 7: `seen[x] = i` stores the current number and its index so a later number can find it quickly.
# Line 8: `return []` returns an empty answer when no valid result was found.

def two_sum(nums, target):
    # Hash map. Time O(n), space O(n).
    seen = {}
    for i, x in enumerate(nums):
        need = target - x
        if need in seen:
            return [seen[need], i]
        seen[x] = i
    return []
```

```python
# Beginner walkthrough:
# Read this section slowly. Each note explains the next meaningful code line, in the same order the code appears.
# Line 1: `def two_sum_sorted(numbers, target):` defines `two_sum_sorted`. The values inside the parentheses are
#        the inputs: `numbers, target`.
# Line 2: `left, right = 0, len(numbers) - 1` places one pointer at the start and one at the end.
# Line 3: `while left < right:` keeps running while the two pointers still describe a real range to check.
# Line 4: `total = numbers[left] + numbers[right]` stores a value in `total` so later lines can use it.
# Line 5: `if total == target:` checks whether this specific case is happening before choosing the next action.
# Line 6: `return [left + 1, right + 1]` returns the answer produced by the previous lines.
# Line 7: `if total < target:` checks whether this specific case is happening before choosing the next action.
# Line 8: `left += 1` moves the left pointer one step right.
# Line 9: `else:` handles the remaining case after the earlier checks failed.
# Line 10: `right -= 1` moves the right pointer one step left.
# Line 11: `return []` returns an empty answer when no valid result was found.

def two_sum_sorted(numbers, target):
    # Sorted array lets us move the pointer that is too small or too large.
    # Time O(n), space O(1). Returns 1-indexed positions.
    left, right = 0, len(numbers) - 1
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]
        if total < target:
            left += 1
        else:
            right -= 1
    return []
```
