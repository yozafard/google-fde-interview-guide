### Two Sum

**Difficulty:** Easy  
**Pattern:** Hash Map  
**Reference:** LeetCode 1  
**Solution to study:** `two_sum`

**Question in plain English:** Given a list of numbers and a target, return the two indexes whose numbers add up to the target.

**Like you are 12:** Pretend each number is a puzzle piece. For every piece, you ask, "What other piece would complete the target?" A dictionary is your memory box for pieces you already saw.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Map` idea and explain what each variable is remembering.

**Best approach:** Use a dictionary when you keep asking, "Have I seen the matching value before?" It lets Python answer that question quickly instead of scanning the list again.

**How to recognize it:** Look for pair lookup, counting, grouping, or "seen before" wording.

**What to say in an interview:** Name the slow brute force idea first, then explain that the dictionary remembers useful past values so each new item only needs one lookup.

**How the solution works:**

1. Create a `seen` dictionary where each number points to its index.
2. For every number, compute the complement `target - x`.
3. If the complement is already in `seen`, the pair is found; otherwise store the current number.

**Beginner checks:** duplicates, empty input, repeated values.


```python
# Beginner walkthrough:
# Step 1: Create a `seen` dictionary where each number points to its index.
# Step 2: For every number, compute the complement `target - x`.
# Step 3: If the complement is already in `seen`, the pair is found; otherwise store the
#         current number.

def two_sum(nums, target):
    # Hash map. Time O(n), space O(n).
    seen = {}
    for i, x in enumerate(nums):
        need = target - x
        if need in seen:
            return [seen[need], i]
        seen[x] = i
    return []
```

### Contains Duplicate

**Difficulty:** Easy  
**Pattern:** Hash Set  
**Reference:** LeetCode 217  
**Solution to study:** `contains_duplicate`

**Question in plain English:** Given a list of numbers, return whether any number appears more than once.

**Like you are 12:** Put every number into a set, like putting stickers into a book where duplicates do not count. If the sticker book is smaller than the original list, something repeated.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Set` idea and explain what each variable is remembering.

**Best approach:** Use a set when you only care whether a value has appeared, not where it appeared or how many times.

**How to recognize it:** Look for duplicate detection, membership checks, or "does this exist?" questions.

**What to say in an interview:** Explain that a set removes repeated scanning because membership checks are fast.

**How the solution works:**

1. Convert the list to a set, which removes duplicates.
2. If the set is shorter than the original list, at least one duplicate existed.

**Beginner checks:** empty input, one item, all same values.


```python
# Beginner walkthrough:
# Step 1: Convert the list to a set, which removes duplicates.
# Step 2: If the set is shorter than the original list, at least one duplicate existed.

def contains_duplicate(nums):
    # Hash set. Time O(n), space O(n).
    return len(nums) != len(set(nums))
```

### Valid Anagram

**Difficulty:** Easy  
**Pattern:** Hash Map/Sorting  
**Reference:** LeetCode 242  
**Solution to study:** `is_anagram`

**Question in plain English:** Given two strings, return whether they use the same letters the same number of times.

**Like you are 12:** Imagine both words are bags of letter tiles. They are anagrams only if both bags contain exactly the same tiles.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Map/Sorting` idea and explain what each variable is remembering.

**Best approach:** Either count characters with a dictionary-like structure or sort both strings so matching strings line up in the same order.

**How to recognize it:** Look for problems asking whether two collections contain the same items with the same frequencies.

**What to say in an interview:** Explain the counting idea: two words match only if every character count matches.

**How the solution works:**

1. Count the characters in both strings with `Counter`.
2. Two strings are anagrams only when every character has the same frequency in both.

**Beginner checks:** duplicates, empty input, repeated values.


```python
# Beginner walkthrough:
# Step 1: Count the characters in both strings with `Counter`.
# Step 2: Two strings are anagrams only when every character has the same frequency in
#         both.

def is_anagram(s, t):
    # Frequency count. Time O(n), space O(1) for lowercase English letters.
    return Counter(s) == Counter(t)
```

### Group Anagrams

**Difficulty:** Medium  
**Pattern:** Hash Map  
**Reference:** LeetCode 49  
**Solution to study:** `group_anagrams`

**Question in plain English:** Given a list of words, group together words that are anagrams of each other.

**Like you are 12:** Sort the letters in each word to make a label. Words with the same label go into the same bucket.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Map` idea and explain what each variable is remembering.

**Best approach:** Use a dictionary when you keep asking, "Have I seen the matching value before?" It lets Python answer that question quickly instead of scanning the list again.

**How to recognize it:** Look for pair lookup, counting, grouping, or "seen before" wording.

**What to say in an interview:** Name the slow brute force idea first, then explain that the dictionary remembers useful past values so each new item only needs one lookup.

**How the solution works:**

1. Sort each word to create a shared key for anagrams.
2. Append each word to the list for that key in a dictionary.
3. Return the grouped dictionary values.

**Beginner checks:** duplicates, empty input, repeated values.


```python
# Beginner walkthrough:
# Step 1: Sort each word to create a shared key for anagrams.
# Step 2: Append each word to the list for that key in a dictionary.
# Step 3: Return the grouped dictionary values.

def group_anagrams(strs):
    # Canonical key by sorted characters. Time O(n * k log k), space O(n * k).
    groups = defaultdict(list)
    for word in strs:
        groups["".join(sorted(word))].append(word)
    return list(groups.values())
```

### Top K Frequent Elements

**Difficulty:** Medium  
**Pattern:** Hash Map + Heap  
**Reference:** LeetCode 347  
**Solution to study:** `top_k_frequent`

**Question in plain English:** Given numbers and k, return the k numbers that appear most often.

**Like you are 12:** Count how many times every number shows up, then pick the most popular numbers.

**What the question is testing:** Whether you can turn the problem into a simple `Hash Map + Heap` idea and explain what each variable is remembering.

**Best approach:** First count how often each value appears, then use a heap/helper to keep or select the largest counts.

**How to recognize it:** Look for "top k", "most frequent", or ranked results based on counts.

**What to say in an interview:** Explain that counting creates the score, and the heap step selects the best k scores.

**How the solution works:**

1. Count each number with `Counter`.
2. Use `heapq.nlargest` to select the keys with the highest counts.
3. This avoids manually sorting by frequency in the solution logic.

**Beginner checks:** duplicates, empty input, repeated values.


```python
# Beginner walkthrough:
# Step 1: Count each number with `Counter`.
# Step 2: Use `heapq.nlargest` to select the keys with the highest counts.
# Step 3: This avoids manually sorting by frequency in the solution logic.

def top_k_frequent(nums, k):
    # Counter + heap helper. Time O(n log k), space O(n).
    count = Counter(nums)
    return heapq.nlargest(k, count.keys(), key=count.get)
```

### Product of Array Except Self

**Difficulty:** Medium  
**Pattern:** Prefix/Suffix  
**Reference:** LeetCode 238  
**Solution to study:** `product_except_self`

**Question in plain English:** Given numbers, return a new list where each position has the product of every other number except itself.

**Like you are 12:** For each spot, multiply everything on its left and everything on its right. Do not use the number sitting in that spot.

**What the question is testing:** Whether you can turn the problem into a simple `Prefix/Suffix` idea and explain what each variable is remembering.

**Best approach:** Precompute information from the left and from the right so each index can be answered without redoing work.

**How to recognize it:** Look for "except self", range products/sums, or questions where each position needs information around it.

**What to say in an interview:** Explain what the left pass knows, what the right pass knows, and how combining them gives the answer.

**How the solution works:**

1. First pass stores the product of everything to the left of each index.
2. Second pass multiplies by the product of everything to the right.
3. No division is needed, so zero values are handled correctly.

**Beginner checks:** empty input, one element, duplicate values.


```python
# Beginner walkthrough:
# Step 1: First pass stores the product of everything to the left of each index.
# Step 2: Second pass multiplies by the product of everything to the right.
# Step 3: No division is needed, so zero values are handled correctly.

def product_except_self(nums):
    # Prefix/suffix without division. Time O(n), space O(1) extra besides output.
    res = [1] * len(nums)
    prefix = 1
    for i, x in enumerate(nums):
        res[i] = prefix
        prefix *= x
    suffix = 1
    for i in range(len(nums) - 1, -1, -1):
        res[i] *= suffix
        suffix *= nums[i]
    return res
```

### Valid Palindrome

**Difficulty:** Easy  
**Pattern:** Two Pointers  
**Reference:** LeetCode 125  
**Solution to study:** `is_palindrome`

**Question in plain English:** Given a string, return whether it reads the same forward and backward after ignoring spaces, punctuation, and capitalization.

**Like you are 12:** Check the word from both ends, like two fingers walking toward the middle. If the fingers see different letters, it is not a palindrome.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers` idea and explain what each variable is remembering.

**Best approach:** Use two indexes that move through the data in a controlled way, often from both ends or as read/write positions.

**How to recognize it:** Look for sorted arrays, pairs, palindromes, in-place edits, or situations where moving one side changes the result predictably.

**What to say in an interview:** Explain what each pointer means and why each move cannot skip a better answer.

**How the solution works:**

1. Place one pointer at the start and one at the end.
2. Skip characters that are not letters or digits.
3. Compare lowercase characters, then move both pointers inward.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Place one pointer at the start and one at the end.
# Step 2: Skip characters that are not letters or digits.
# Step 3: Compare lowercase characters, then move both pointers inward.

def is_palindrome(s):
    # Opposite-end pointers. Time O(n), space O(1).
    left, right = 0, len(s) - 1
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True
```

### Two Sum II (sorted)

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 167  
**Solution to study:** `two_sum_sorted`

**Question in plain English:** Given a sorted list and a target, return the two positions whose numbers add up to the target.

**Like you are 12:** Use one finger on the small end and one finger on the big end. If the sum is too small, move the small finger up. If too big, move the big finger down.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers` idea and explain what each variable is remembering.

**Best approach:** Use two indexes that move through the data in a controlled way, often from both ends or as read/write positions.

**How to recognize it:** Look for sorted arrays, pairs, palindromes, in-place edits, or situations where moving one side changes the result predictably.

**What to say in an interview:** Explain what each pointer means and why each move cannot skip a better answer.

**How the solution works:**

1. Use the sorted order to avoid a hash map.
2. If the two-pointer sum is too small, move `left` rightward.
3. If it is too large, move `right` leftward.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Use the sorted order to avoid a hash map.
# Step 2: If the two-pointer sum is too small, move `left` rightward.
# Step 3: If it is too large, move `right` leftward.

def two_sum_sorted(numbers, target):
    # Sorted array lets us move the pointer that is too small or too large.
    # Time O(n), space O(1). Returns 1-indexed positions.
    left, right = 0, len(numbers) - 1
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]
        if total < target:
            left += 1
        else:
            right -= 1
    return []
```

### 3Sum

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 15  
**Solution to study:** `three_sum`

**Question in plain English:** Given numbers, return all unique groups of three numbers that add up to zero.

**Like you are 12:** Pick one number first, then use two fingers to find two more numbers that balance it back to zero.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers` idea and explain what each variable is remembering.

**Best approach:** Use two indexes that move through the data in a controlled way, often from both ends or as read/write positions.

**How to recognize it:** Look for sorted arrays, pairs, palindromes, in-place edits, or situations where moving one side changes the result predictably.

**What to say in an interview:** Explain what each pointer means and why each move cannot skip a better answer.

**How the solution works:**

1. Sort first so duplicates can be skipped and two pointers can be used.
2. Fix one number at index `i`, then search for the other two numbers with `left` and `right`.
3. After finding a triplet, skip duplicate left and right values.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Sort first so duplicates can be skipped and two pointers can be used.
# Step 2: Fix one number at index `i`, then search for the other two numbers with `left`
#         and `right`.
# Step 3: After finding a triplet, skip duplicate left and right values.

def three_sum(nums):
    # Sort + two pointers. Time O(n^2), space O(1) extra besides output.
    nums.sort()
    res = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            elif total < 0:
                left += 1
            else:
                right -= 1
    return res
```

### Container With Most Water

**Difficulty:** Medium  
**Pattern:** Two Pointers  
**Reference:** LeetCode 11  
**Solution to study:** `max_area`

**Question in plain English:** Given wall heights, find the two walls that can hold the most water.

**Like you are 12:** Two walls make a bucket. Start with the widest bucket, then move the shorter wall because the shorter wall is what limits the water.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers` idea and explain what each variable is remembering.

**Best approach:** Use two indexes that move through the data in a controlled way, often from both ends or as read/write positions.

**How to recognize it:** Look for sorted arrays, pairs, palindromes, in-place edits, or situations where moving one side changes the result predictably.

**What to say in an interview:** Explain what each pointer means and why each move cannot skip a better answer.

**How the solution works:**

1. Start with the widest possible container.
2. Compute the area using the shorter wall.
3. Move the shorter wall inward because moving the taller wall cannot improve the limiting height.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Start with the widest possible container.
# Step 2: Compute the area using the shorter wall.
# Step 3: Move the shorter wall inward because moving the taller wall cannot improve the
#         limiting height.

def max_area(height):
    # Move the shorter wall because width always shrinks. Time O(n), space O(1).
    left, right = 0, len(height) - 1
    best = 0
    while left < right:
        best = max(best, min(height[left], height[right]) * (right - left))
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    return best
```

### Trapping Rain Water

**Difficulty:** Hard  
**Pattern:** Two Pointers/DP  
**Reference:** LeetCode 42  
**Solution to study:** `trap`

**Question in plain English:** Given wall heights, compute how much rain water can be trapped between them.

**Like you are 12:** Water above a spot depends on the tallest wall on the left and the tallest wall on the right. The shorter side decides how high water can sit.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers/DP` idea and explain what each variable is remembering.

**Best approach:** Track useful information from both sides so you do not recompute it for every position.

**How to recognize it:** Look for problems where the answer at a position depends on the best boundary to the left and right.

**What to say in an interview:** Explain which side currently limits the answer, then move that side because it is the only side that can improve.

**How the solution works:**

1. Track the best wall seen from the left and from the right.
2. Move the side with the smaller current height because that side limits trapped water.
3. Add the difference between the side max and current height.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Track the best wall seen from the left and from the right.
# Step 2: Move the side with the smaller current height because that side limits trapped
#         water.
# Step 3: Add the difference between the side max and current height.

def trap(height):
    # Two pointers with left/right max. Time O(n), space O(1).
    left, right = 0, len(height) - 1
    left_max = right_max = water = 0
    while left < right:
        if height[left] < height[right]:
            left_max = max(left_max, height[left])
            water += left_max - height[left]
            left += 1
        else:
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1
    return water
```

### Remove Duplicates from Sorted Array

**Difficulty:** Easy  
**Pattern:** Two Pointers  
**Reference:** LeetCode 26  
**Solution to study:** `remove_duplicates`

**Question in plain English:** Given a sorted list, remove duplicates in-place and return how many unique numbers remain.

**Like you are 12:** Use one finger to read every number and another finger to write only the first copy of each new number.

**What the question is testing:** Whether you can turn the problem into a simple `Two Pointers` idea and explain what each variable is remembering.

**Best approach:** Use two indexes that move through the data in a controlled way, often from both ends or as read/write positions.

**How to recognize it:** Look for sorted arrays, pairs, palindromes, in-place edits, or situations where moving one side changes the result predictably.

**What to say in an interview:** Explain what each pointer means and why each move cannot skip a better answer.

**How the solution works:**

1. Use `read` to scan the array and `write` to place unique values.
2. Because the array is sorted, a new value differs from the previous value.
3. Return `write`, the length of the unique prefix.

**Beginner checks:** duplicates, empty input, pointers crossing.


```python
# Beginner walkthrough:
# Step 1: Use `read` to scan the array and `write` to place unique values.
# Step 2: Because the array is sorted, a new value differs from the previous value.
# Step 3: Return `write`, the length of the unique prefix.

def remove_duplicates(nums):
    # Write pointer keeps the next unique position. Time O(n), space O(1).
    if not nums:
        return 0
    write = 1
    for read in range(1, len(nums)):
        if nums[read] != nums[read - 1]:
            nums[write] = nums[read]
            write += 1
    return write
```

### Best Time to Buy and Sell Stock

**Difficulty:** Easy  
**Pattern:** Sliding Window  
**Reference:** LeetCode 121  
**Solution to study:** `max_profit`

**Question in plain English:** Given daily prices, return the biggest profit from buying once and selling later.

**Like you are 12:** Keep track of the cheapest price you have seen. Each day, ask, "If I sold today, how much would I make?"

**What the question is testing:** Whether you can turn the problem into a simple `Sliding Window` idea and explain what each variable is remembering.

**Best approach:** Keep a contiguous chunk of the array/string, expand it to include more items, and shrink it when it breaks the rule.

**How to recognize it:** Look for substring/subarray wording, longest/shortest ranges, fixed length `k`, or "at most k" constraints.

**What to say in an interview:** Explain what makes the current window valid and when the left or right edge moves.

**How the solution works:**

1. Track the lowest buy price seen so far.
2. For each price, compute the profit if sold today.
3. Keep the maximum profit across all days.

**Beginner checks:** empty string, k is 0, repeated characters.


```python
# Beginner walkthrough:
# Step 1: Track the lowest buy price seen so far.
# Step 2: For each price, compute the profit if sold today.
# Step 3: Keep the maximum profit across all days.

def max_profit(prices):
    # Track lowest buy so far. Time O(n), space O(1).
    best = 0
    buy = float("inf")
    for price in prices:
        buy = min(buy, price)
        best = max(best, price - buy)
    return best
```

### Maximum Subarray

**Difficulty:** Medium  
**Pattern:** Kadane's  
**Reference:** LeetCode 53  
**Solution to study:** `max_sub_array`

**Question in plain English:** Given numbers, return the biggest possible sum from one continuous section.

**Like you are 12:** As you walk across the numbers, decide whether your current group is helping or hurting. If it hurts, start fresh at the current number.

**What the question is testing:** Whether you can turn the problem into a simple `Kadane's` idea and explain what each variable is remembering.

**Best approach:** At each number, decide whether it is better to continue the previous subarray or start a new one here.

**How to recognize it:** Look for maximum subarray or best contiguous sum/product wording.

**What to say in an interview:** Explain that the running value means "best sum ending exactly here," while the answer stores the best seen anywhere.

**How the solution works:**

1. `curr` means the best subarray sum ending at the current index.
2. At each number, choose between starting fresh or extending the previous subarray.
3. `best` stores the best sum seen anywhere.

**Beginner checks:** all negative numbers, one element.


```python
# Beginner walkthrough:
# Step 1: `curr` means the best subarray sum ending at the current index.
# Step 2: At each number, choose between starting fresh or extending the previous
#         subarray.
# Step 3: `best` stores the best sum seen anywhere.

def max_sub_array(nums):
    # Kadane's algorithm. Time O(n), space O(1).
    best = curr = nums[0]
    for x in nums[1:]:
        curr = max(x, curr + x)
        best = max(best, curr)
    return best
```

### Longest Substring Without Repeating Characters

**Difficulty:** Medium  
**Pattern:** Sliding Window  
**Reference:** LeetCode 3  
**Solution to study:** `length_of_longest_substring`

**Question in plain English:** Given a string, return the length of the longest continuous part with no repeated characters.

**Like you are 12:** Drag a window across the string. If a letter repeats inside the window, move the left side forward until every letter is unique again.

**What the question is testing:** Whether you can turn the problem into a simple `Sliding Window` idea and explain what each variable is remembering.

**Best approach:** Keep a contiguous chunk of the array/string, expand it to include more items, and shrink it when it breaks the rule.

**How to recognize it:** Look for substring/subarray wording, longest/shortest ranges, fixed length `k`, or "at most k" constraints.

**What to say in an interview:** Explain what makes the current window valid and when the left or right edge moves.

**How the solution works:**

1. Keep a window with no repeated characters.
2. Store each character's last seen index.
3. When a repeat appears inside the current window, move `left` past the previous occurrence.

**Beginner checks:** empty string, k is 0, repeated characters.


```python
# Beginner walkthrough:
# Step 1: Keep a window with no repeated characters.
# Step 2: Store each character's last seen index.
# Step 3: When a repeat appears inside the current window, move `left` past the previous
#         occurrence.

def length_of_longest_substring(s):
    # Variable window with last-seen positions. Time O(n), space O(k).
    last = {}
    left = best = 0
    for right, ch in enumerate(s):
        if ch in last and last[ch] >= left:
            left = last[ch] + 1
        last[ch] = right
        best = max(best, right - left + 1)
    return best
```

### Longest Repeating Character Replacement

**Difficulty:** Medium  
**Pattern:** Sliding Window  
**Reference:** LeetCode 424  
**Solution to study:** `character_replacement`

**Question in plain English:** Given a string and k changes, find the longest substring that can become all one repeated character.

**Like you are 12:** Keep a window of letters. If too many letters would need changing, shrink the window until it is possible again.

**What the question is testing:** Whether you can turn the problem into a simple `Sliding Window` idea and explain what each variable is remembering.

**Best approach:** Keep a contiguous chunk of the array/string, expand it to include more items, and shrink it when it breaks the rule.

**How to recognize it:** Look for substring/subarray wording, longest/shortest ranges, fixed length `k`, or "at most k" constraints.

**What to say in an interview:** Explain what makes the current window valid and when the left or right edge moves.

**How the solution works:**

1. Track character counts inside the current window.
2. The window is valid when `window_size - most_common_count <= k`.
3. Shrink from the left only when more than `k` replacements are needed.

**Beginner checks:** empty string, k is 0, repeated characters.


```python
# Beginner walkthrough:
# Step 1: Track character counts inside the current window.
# Step 2: The window is valid when `window_size - most_common_count <= k`.
# Step 3: Shrink from the left only when more than `k` replacements are needed.

def character_replacement(s, k):
    # Keep window where changes_needed <= k. Time O(n), space O(1).
    count = defaultdict(int)
    left = best = max_count = 0
    for right, ch in enumerate(s):
        count[ch] += 1
        max_count = max(max_count, count[ch])
        while (right - left + 1) - max_count > k:
            count[s[left]] -= 1
            left += 1
        best = max(best, right - left + 1)
    return best
```

### Minimum Window Substring

**Difficulty:** Hard  
**Pattern:** Sliding Window

**Question in plain English:** Given strings s and t, find the shortest part of s that contains every character from t.

**Like you are 12:** You are trying to cover all required letters with the smallest possible blanket. Make the blanket bigger until it covers them, then shrink it as much as you can.

**What the question is testing:** Whether you can turn the problem into a simple `Sliding Window` idea and explain what each variable is remembering.

**Best approach:** Keep a contiguous chunk of the array/string, expand it to include more items, and shrink it when it breaks the rule.

**How to recognize it:** Look for substring/subarray wording, longest/shortest ranges, fixed length `k`, or "at most k" constraints.

**What to say in an interview:** Explain what makes the current window valid and when the left or right edge moves.

**How the solution works:**

1. Count the characters required by `t`.
2. Expand right until the current window covers all required characters.
3. Then shrink from the left to make the valid window as small as possible.

**Beginner checks:** empty string, k is 0, repeated characters.


```python
# Beginner walkthrough:
# Step 1: Count the characters required by `t`.
# Step 2: Expand right until the current window covers all required characters.
# Step 3: Then shrink from the left to make the valid window as small as possible.

def min_window(s, t):
    # Expand until valid, shrink to minimal. Time O(n), space O(k).
    need = Counter(t)
    missing = len(t)
    left = start = end = 0
    for right, ch in enumerate(s, 1):
        if need[ch] > 0:
            missing -= 1
        need[ch] -= 1
        if missing == 0:
            while left < right and need[s[left]] < 0:
                need[s[left]] += 1
                left += 1
            if end == 0 or right - left < end - start:
                start, end = left, right
            need[s[left]] += 1
            missing += 1
            left += 1
    return s[start:end]
```

### Sliding Window Maximum

**Difficulty:** Hard  
**Pattern:** Deque  
**Reference:** LeetCode 239  
**Solution to study:** `max_sliding_window`

**Question in plain English:** Given numbers and a window size k, return the biggest number in each sliding window.

**Like you are 12:** As the window moves, keep a line of useful candidates for biggest number. Throw away numbers that are too old or clearly smaller than a newer number.

**What the question is testing:** Whether you can turn the problem into a simple `Deque` idea and explain what each variable is remembering.

**Best approach:** Use a double-ended queue to keep only candidates that can still become the answer for the current window.

**How to recognize it:** Look for sliding-window maximum/minimum where old items expire and weaker items can be removed.

**What to say in an interview:** Explain why the front of the deque is always the current best candidate.

**How the solution works:**

1. Store indexes in a deque, not values, so stale indexes can be removed.
2. Keep the deque decreasing by value, so the front is always the max.
3. Once a full window is formed, append `nums[q[0]]`.

**Beginner checks:** stale indexes, duplicate values.


```python
# Beginner walkthrough:
# Step 1: Store indexes in a deque, not values, so stale indexes can be removed.
# Step 2: Keep the deque decreasing by value, so the front is always the max.
# Step 3: Once a full window is formed, append `nums[q[0]]`.

def max_sliding_window(nums, k):
    # Monotonic deque stores candidate indexes. Time O(n), space O(k).
    q = deque()
    res = []
    for i, x in enumerate(nums):
        while q and q[0] <= i - k:
            q.popleft()
        while q and nums[q[-1]] <= x:
            q.pop()
        q.append(i)
        if i >= k - 1:
            res.append(nums[q[0]])
    return res
```

### Min Stack

**Pattern:** Stack  
**Reference:** LeetCode 155  
**Solution to study:** `MinStack`

**Question in plain English:** Design a stack that can also return the smallest item in constant time.

**Like you are 12:** Keep the normal stack, plus a helper stack that remembers the smallest value at each moment.

**What the question is testing:** Whether you can turn the problem into a simple `Stack` idea and explain what each variable is remembering.

**Best approach:** Use a stack when the most recent unfinished item must be handled first.

**How to recognize it:** Look for matching brackets, undo-like behavior, nested structure, or postfix expressions.

**What to say in an interview:** Explain what gets pushed, what causes a pop, and what the stack represents after each step.

**How the solution works:**

1. Store each value together with the minimum value at that moment.
2. When popping, the previous minimum is restored automatically.
3. `getMin` simply reads the saved minimum at the stack top.

**Beginner checks:** empty stack, invalid order, duplicate heights.


```python
# Beginner walkthrough:
# Step 1: Store each value together with the minimum value at that moment.
# Step 2: When popping, the previous minimum is restored automatically.
# Step 3: `getMin` simply reads the saved minimum at the stack top.

class MinStack:
    # Store current min beside each value. All operations O(1).
    def __init__(self):
        self.stack = []

    def push(self, val):
        curr_min = val if not self.stack else min(val, self.stack[-1][1])
        self.stack.append((val, curr_min))

    def pop(self):
        self.stack.pop()

    def top(self):
        return self.stack[-1][0]

    def getMin(self):
        return self.stack[-1][1]
```

**How to recognize it:** Look for matching brackets, undo-like behavior, nested structure, or postfix expressions.

**What to say in an interview:** Explain what gets pushed, what causes a pop, and what the stack represents after each step.

**How the solution works:**

3. Apply the operation and push the result back.

**Beginner checks:** empty stack, invalid order, duplicate heights.


```python
# Beginner walkthrough:
# Step 1: Push numbers onto a stack.
# Step 2: For an operator, pop the last two numbers in operand order.
# Step 3: Apply the operation and push the result back.

def eval_rpn(tokens):
    # Stack operands. Time O(n), space O(n).
    stack = []
    for token in tokens:
        if token not in "+-*/":
            stack.append(int(token))
            continue
        b, a = stack.pop(), stack.pop()
        if token == "+":
            stack.append(a + b)
        elif token == "-":
            stack.append(a - b)
        elif token == "*":
            stack.append(a * b)
        else:
            stack.append(int(a / b))
    return stack[-1]
```

**How to recognize it:** Look for next greater/smaller element, temperatures, spans, or histogram rectangles.

**What to say in an interview:** Explain what unresolved items are waiting in the stack and what event resolves them.

**How the solution works:**

1. Keep a stack of indexes whose warmer day has not been found yet.
2. When today is warmer than the temperature at the stack top, resolve that older day.

```python
# Beginner walkthrough:
# Step 1: Keep a stack of indexes whose warmer day has not been found yet.
# Step 2: When today is warmer than the temperature at the stack top, resolve that older
#         day.
# Step 3: The difference in indexes is the wait time.

def daily_temperatures(temperatures):
    # Monotonic decreasing stack of indexes. Time O(n), space O(n).
    res = [0] * len(temperatures)
    stack = []
    for i, temp in enumerate(temperatures):
        while stack and temp > temperatures[stack[-1]]:
            j = stack.pop()
            res[j] = i - j
        stack.append(i)
    return res
```

### Largest Rectangle in Histogram

**What the question is testing:** Whether you can turn the problem into a simple `Monotonic Stack` idea and explain what each variable is remembering.

**How to recognize it:** Look for next greater/smaller element, temperatures, spans, or histogram rectangles.

**What to say in an interview:** Explain what unresolved items are waiting in the stack and what event resolves them.

**How the solution works:**

1. Keep an increasing stack of bar heights with their earliest start index.
2. When a shorter bar appears, rectangles using taller bars must end here.
3. A sentinel zero at the end forces all remaining bars to be processed.

**Beginner checks:** empty stack, invalid order, duplicate heights.


```python
# Beginner walkthrough:
# Step 1: Keep an increasing stack of bar heights with their earliest start index.
# Step 2: When a shorter bar appears, rectangles using taller bars must end here.
# Step 3: A sentinel zero at the end forces all remaining bars to be processed.

def largest_rectangle_area(heights):
    # Monotonic stack. Time O(n), space O(n).
    stack = []
    best = 0
    for i, h in enumerate(heights + [0]):
        start = i
        while stack and stack[-1][1] > h:
            index, height = stack.pop()
            best = max(best, height * (i - index))
            start = index
        stack.append((start, h))
    return best
```

### Binary Search

**Difficulty:** Easy  
**Pattern:** Standard  
**Reference:** LeetCode 704  
**Solution to study:** `search`

**Question in plain English:** Given a sorted list and a target, return the target index or -1 if missing.

**Like you are 12:** Open the book in the middle. If the target is smaller, ignore the right half. If bigger, ignore the left half.

**What the question is testing:** Whether you can turn the problem into a simple `Standard` idea and explain what each variable is remembering.

**Best approach:** Use binary search on a sorted array by repeatedly cutting away the half that cannot contain the target.

**How to recognize it:** Look for sorted input and direct target lookup.

**What to say in an interview:** Explain the current search range and why each comparison safely removes half of it.

**How the solution works:**

```python
# Beginner walkthrough:
# Step 1: Maintain a search range from `left` to `right`.
# Step 2: Compare the middle value with the target.
# Step 3: Discard the half that cannot contain the target.

def binary_search(nums, target):
    # Standard binary search. Time O(log n), space O(1).
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

**Like you are 12:** Use binary search twice: once to find the left edge of the target group and once to find the right edge.

**What the question is testing:** Whether you can turn the problem into a simple `Boundaries` idea and explain what each variable is remembering.

**Best approach:** Run boundary-style binary search twice: once for the first valid position and once for the position after the last match.

**How to recognize it:** Look for first/last occurrence or count of a target in sorted data.

**What to say in an interview:** Explain that the second boundary makes the ending index easy to compute.

**How the solution works:**

1. Find the first index where target could appear using lower bound.
2. Find the first index greater than target using lower bound on `target + 1`.
3. The last occurrence is one index before that second boundary.

**Beginner checks:** empty input, one element, duplicate values.


```python
# Beginner walkthrough:
# Step 1: Find the first index where target could appear using lower bound.
# Step 2: Find the first index greater than target using lower bound on `target + 1`.
# Step 3: The last occurrence is one index before that second boundary.

def search_range(nums, target):
    # Two lower-bound searches. Time O(log n), space O(1).
    def lower_bound(x):
        left, right = 0, len(nums)
        while left < right:
            mid = (left + right) // 2
            if nums[mid] < x:
                left = mid + 1
            else:
                right = mid
        return left

    start = lower_bound(target)
    if start == len(nums) or nums[start] != target:
        return [-1, -1]
    return [start, lower_bound(target + 1) - 1]
```

```python
# Beginner walkthrough:
# Step 1: Treat the matrix as a flattened sorted array.
# Step 2: Convert a flat index back to row and column with division and modulo.
# Step 3: Run normal binary search over the virtual array.

def search_matrix(matrix, target):
    # Treat matrix as flattened sorted array. Time O(log(mn)), space O(1).
    if not matrix or not matrix[0]:
        return False
    rows, cols = len(matrix), len(matrix[0])
    left, right = 0, rows * cols - 1
    while left <= right:
        mid = (left + right) // 2
        value = matrix[mid // cols][mid % cols]
        if value == target:
            return True
        if value < target:
            left = mid + 1
        else:
            right = mid - 1
    return False
```

```python
# Beginner walkthrough:
# Step 1: First detect a meeting point with fast and slow pointers.
# Step 2: Reset one pointer to the head.
# Step 3: Move both one step at a time; their meeting point is the cycle start.

def detect_cycle(head):
    # After meeting, reset one pointer to head. Time O(n), space O(1).
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow is fast:
            slow = head
            while slow is not fast:
                slow = slow.next
                fast = fast.next
            return slow
    return None
```

### Binary Tree Level Order Traversal

**Difficulty:** Medium  
**Pattern:** BFS  
**Reference:** LeetCode 102  
**Solution to study:** `level_order`

**Question in plain English:** Given a binary tree, return its values level by level.

**Like you are 12:** Use a queue like a waiting line. Process all nodes on the current floor before moving to the next floor.

**What the question is testing:** Whether you can turn the problem into a simple `BFS` idea and explain what each variable is remembering.

**Best approach:** Explore level by level with a queue, which is especially useful for shortest paths with equal-cost steps.

**How to recognize it:** Look for level order traversal, shortest path in an unweighted graph, or spreading outward one step at a time.

**What to say in an interview:** Explain what one queue level represents and when nodes are marked visited.

**How the solution works:**

1. Use a queue for BFS.
2. For each level, process exactly the number of nodes currently in the queue.
3. Append children to the queue for the next level.

**Beginner checks:** start equals end, empty input, disconnected graph.


```python
# Beginner walkthrough:
# Step 1: Use a queue for BFS.
# Step 2: For each level, process exactly the number of nodes currently in the queue.
# Step 3: Append children to the queue for the next level.

def level_order(root):
    # BFS by levels. Time O(n), space O(w).
    if not root:
        return []
    q = deque([root])
    res = []
    while q:
        level = []
        for _ in range(len(q)):
            node = q.popleft()
            level.append(node.val)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        res.append(level)
    return res
```

**How the solution works:**

2. A node must be strictly inside its allowed range.
3. The left child tightens the upper bound; the right child tightens the lower bound.

### Construct Binary Tree from Preorder and Inorder

**Question in plain English:** Given preorder and inorder traversals, rebuild the original tree.

**Like you are 12:** Preorder tells you the root first. Inorder tells you what belongs on the left and right of that root.

**What the question is testing:** Whether you can turn the problem into a simple `Recursion` idea and explain what each variable is remembering.

**Best approach:** Use the `Recursion` pattern to store the small pieces of information that prevent repeated work.

**How to recognize it:** Look for wording that matches this pattern in the study guide.

**What to say in an interview:** Explain the brute force idea, then explain what repeated work the pattern removes.

**How the solution works:**

1. Preorder gives the next root.
2. Inorder tells which values belong to the left and right subtrees.
3. A hashmap gives each inorder index in O(1).

**Beginner checks:** empty input, one element, duplicate values.


```python
# Beginner walkthrough:
# Step 1: Preorder gives the next root.
# Step 2: Inorder tells which values belong to the left and right subtrees.
# Step 3: A hashmap gives each inorder index in O(1).

def build_tree(preorder, inorder):
    # Preorder root + inorder split. Time O(n), space O(n).
    index = {value: i for i, value in enumerate(inorder)}
    pre_i = 0

    def helper(left, right):
        nonlocal pre_i
        if left > right:
            return None
        root_val = preorder[pre_i]
        pre_i += 1
        root = TreeNode(root_val)
        mid = index[root_val]
        root.left = helper(left, mid - 1)
        root.right = helper(mid + 1, right)
        return root

    return helper(0, len(inorder) - 1)
```

### Serialize and Deserialize Binary Tree

**Difficulty:** Hard  
**Pattern:** BFS/DFS  
**Reference:** LeetCode 297  
**Solution to study:** `Codec`

**Question in plain English:** Turn a binary tree into a string and then rebuild the same tree from that string.

**Like you are 12:** Write down every node, including empty spots, so you can later replay the notes and rebuild the exact tree.

**What the question is testing:** Whether you can turn the problem into a simple `BFS/DFS` idea and explain what each variable is remembering.

**Best approach:** Use traversal to visit every needed node in a consistent order; the key is preserving enough information to rebuild or compare structure.

**How to recognize it:** Look for tree/graph traversal, serialization, cloning, or full-structure processing.

**What to say in an interview:** Explain what information is saved for each node and why missing nodes may also need markers.

**How the solution works:**

1. Serialization stores values in BFS order with `#` for missing children.
2. The null markers preserve the exact tree shape.
3. Deserialization rebuilds children in the same BFS order.

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


```python
# Beginner walkthrough:
# Step 1: Serialization stores values in BFS order with `#` for missing children.
# Step 2: The null markers preserve the exact tree shape.
# Step 3: Deserialization rebuilds children in the same BFS order.

class Codec:
    # BFS serialize/deserialize. Time O(n), space O(n).
    def serialize(self, root):
        if not root:
            return ""
        q = deque([root])
        vals = []
        while q:
            node = q.popleft()
            if node:
                vals.append(str(node.val))
                q.append(node.left)
                q.append(node.right)
            else:
                vals.append("#")
        return ",".join(vals)

    def deserialize(self, data):
        if not data:
            return None
        vals = data.split(",")
        root = TreeNode(int(vals[0]))
        q = deque([root])
        i = 1
        while q:
            node = q.popleft()
            if vals[i] != "#":
                node.left = TreeNode(int(vals[i]))
                q.append(node.left)
            i += 1
            if vals[i] != "#":
                node.right = TreeNode(int(vals[i]))
                q.append(node.right)
            i += 1
        return root
```

**Solution to study:** `num_islands`

**Question in plain English:** Given a grid of land and water, count how many separate islands exist.

**Like you are 12:** When you find land, explore all connected land and mark it visited. That whole explored blob is one island.

**What the question is testing:** Whether you can turn the problem into a simple `DFS/BFS` idea and explain what each variable is remembering.

**Best approach:** Use traversal to visit all connected cells/nodes from a starting point before counting another group.

**How to recognize it:** Look for grids, islands, components, or connected regions.

**What to say in an interview:** Explain how visited cells are marked so the same region is not counted twice.

**How the solution works:**

1. Scan every grid cell.
2. When land is found, count one island and DFS from that cell.
3. The DFS marks connected land as water so it is not counted again.

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


```python
# Beginner walkthrough:
# Step 1: Scan every grid cell.
# Step 2: When land is found, count one island and DFS from that cell.
# Step 3: The DFS marks connected land as water so it is not counted again.

def num_islands(grid):
    # DFS flood fill. Time O(mn), space O(mn) worst case.
    if not grid:
        return 0
    rows, cols = len(grid), len(grid[0])
    count = 0

    def dfs(r, c):
        if r < 0 or c < 0 or r == rows or c == cols or grid[r][c] != "1":
            return
        grid[r][c] = "0"
        dfs(r + 1, c)
        dfs(r - 1, c)
        dfs(r, c + 1)
        dfs(r, c - 1)

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1":
                count += 1
                dfs(r, c)
    return count
```

**Question in plain English:** Given a graph node, create a deep copy of the whole connected graph.

**Like you are 12:** Make a copy of each node once, then connect the copied nodes the same way the original nodes are connected.

**How to recognize it:** Look for graph cloning or traversal with cycles.

**What to say in an interview:** Explain that the dictionary prevents infinite loops and lets shared neighbors stay shared.

**How the solution works:**

1. Use a dictionary from original node to cloned node.
2. DFS creates a clone the first time a node is seen.
3. Neighbors are cloned recursively, which handles cycles safely.

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


```python
# Beginner walkthrough:
# Step 1: Use a dictionary from original node to cloned node.
# Step 2: DFS creates a clone the first time a node is seen.
# Step 3: Neighbors are cloned recursively, which handles cycles safely.

def clone_graph(node):
    # DFS with old-node to new-node map. Time O(V + E), space O(V).
    if not node:
        return None
    clones = {}

    def dfs(curr):
        if curr in clones:
            return clones[curr]
        copy = Node(curr.val)
        clones[curr] = copy
        copy.neighbors = [dfs(nei) for nei in curr.neighbors]
        return copy

    return dfs(node)
```

### Pacific Atlantic Water Flow

**Difficulty:** Medium  
**Pattern:** Multi-source BFS  
**Reference:** LeetCode 417  
**Solution to study:** `pacific_atlantic`

**Question in plain English:** Given heights, return cells where water can flow to both oceans.

**Like you are 12:** Instead of starting from every cell, start from the oceans and move uphill. Cells reached by both oceans are answers.

**What the question is testing:** Whether you can turn the problem into a simple `Multi-source BFS` idea and explain what each variable is remembering.

**Best approach:** Start traversal from many useful starting cells at once instead of repeating the same search from every cell.

**How to recognize it:** Look for grid flow/reachability problems with several borders or sources.

**What to say in an interview:** Explain why reversing the direction of the search makes the problem simpler.

**How the solution works:**

1. Start from ocean borders instead of every cell.
2. DFS inward only to cells of equal or greater height.
3. Cells reached from both ocean searches are the answer.

**Beginner checks:** start equals end, empty input, disconnected graph.


```python
# Beginner walkthrough:
# Step 1: Start from ocean borders instead of every cell.
# Step 2: DFS inward only to cells of equal or greater height.
# Step 3: Cells reached from both ocean searches are the answer.

def pacific_atlantic(heights):
    # Reverse flow from oceans inward. Time O(mn), space O(mn).
    rows, cols = len(heights), len(heights[0])
    pac, atl = set(), set()

    def dfs(r, c, seen, prev):
        if (r, c) in seen or r < 0 or c < 0 or r == rows or c == cols or heights[r][c] < prev:
            return
        seen.add((r, c))
        for dr, dc in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            dfs(r + dr, c + dc, seen, heights[r][c])

    for c in range(cols):
        dfs(0, c, pac, heights[0][c])
        dfs(rows - 1, c, atl, heights[rows - 1][c])
    for r in range(rows):
        dfs(r, 0, pac, heights[r][0])
        dfs(r, cols - 1, atl, heights[r][cols - 1])
    return list(pac & atl)
```

**What the question is testing:** Whether you can turn the problem into a simple `Topological Sort` idea and explain what each variable is remembering.

**Best approach:** Process items only after their prerequisites are satisfied.

**How to recognize it:** Look for courses, tasks, dependencies, ordering, or cycle detection in prerequisites.

**What to say in an interview:** Explain indegree as "how many prerequisites are still blocking this item."

**How the solution works:**

1. Build a graph from prerequisite to course.
2. Track how many prerequisites each course still has.
3. Repeatedly take zero-prerequisite courses; if all are taken, no cycle exists.

**Beginner checks:** cycle, no prerequisites, multiple valid orders.


```python
# Beginner walkthrough:
# Step 1: Build a graph from prerequisite to course.
# Step 2: Track how many prerequisites each course still has.
# Step 3: Repeatedly take zero-prerequisite courses; if all are taken, no cycle exists.

def can_finish(num_courses, prerequisites):
    # Kahn topological sort. Time O(V + E), space O(V + E).
    graph = defaultdict(list)
    indegree = [0] * num_courses
    for course, prereq in prerequisites:
        graph[prereq].append(course)
        indegree[course] += 1
    q = deque([i for i, deg in enumerate(indegree) if deg == 0])
    taken = 0
    while q:
        node = q.popleft()
        taken += 1
        for nei in graph[node]:
            indegree[nei] -= 1
            if indegree[nei] == 0:
                q.append(nei)
    return taken == num_courses
```

### Number of Connected Components

**Difficulty:** Medium  
**Pattern:** Union Find/DFS  
**Reference:** LeetCode 323  
**Solution to study:** `count_components`

**Question in plain English:** Given n nodes and edges, count how many separate groups exist.

**Like you are 12:** Each time you find an unseen node, explore everything connected to it. That exploration is one group.

**What the question is testing:** Whether you can turn the problem into a simple `Union Find/DFS` idea and explain what each variable is remembering.

**Best approach:** Group connected nodes together, either by unioning edges or by DFS from each unvisited node.

**How to recognize it:** Look for connected components, networks, or "how many groups?" questions.

**What to say in an interview:** Explain that each successful union reduces the number of separate groups by one.

**How the solution works:**

1. Start with every node as its own component.
2. Union the two endpoints of each edge.
3. Each successful union reduces the component count by one.

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


```python
# Beginner walkthrough:
# Step 1: Start with every node as its own component.
# Step 2: Union the two endpoints of each edge.
# Step 3: Each successful union reduces the component count by one.

def count_components(n, edges):
    # Union find. Time near O(V + E), space O(V).
    parent = list(range(n))
    rank = [0] * n

    def find(x):
        while x != parent[x]:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x

    def union(a, b):
        ra, rb = find(a), find(b)
        if ra == rb:
            return False
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra] += 1
        return True

    components = n
    for a, b in edges:
        if union(a, b):
            components -= 1
    return components
```

### Word Ladder

**Difficulty:** Hard  
**Pattern:** BFS  
**Reference:** LeetCode 127  
**Solution to study:** `ladder_length`

**Question in plain English:** Given a start word, end word, and dictionary, find the shortest number of one-letter changes to reach the end.

**Like you are 12:** Treat each word as a stop on a map. You can travel to another stop if only one letter changes, and BFS finds the shortest trip.

**What the question is testing:** Whether you can turn the problem into a simple `BFS` idea and explain what each variable is remembering.

**Best approach:** Explore level by level with a queue, which is especially useful for shortest paths with equal-cost steps.

**How to recognize it:** Look for level order traversal, shortest path in an unweighted graph, or spreading outward one step at a time.

**What to say in an interview:** Explain what one queue level represents and when nodes are marked visited.

**How the solution works:**

1. Create wildcard buckets like `h*t` to connect one-letter-different words.
2. Run BFS from the begin word.
3. The first time the end word is reached is the shortest transformation length.

**Beginner checks:** start equals end, empty input, disconnected graph.


```python
# Beginner walkthrough:
# Step 1: Create wildcard buckets like `h*t` to connect one-letter-different words.
# Step 2: Run BFS from the begin word.
# Step 3: The first time the end word is reached is the shortest transformation length.

def ladder_length(begin_word, end_word, word_list):
    # BFS over wildcard buckets. Time O(n * m^2), space O(n * m).
    words = set(word_list)
    if end_word not in words:
        return 0
    buckets = defaultdict(list)
    for word in words | {begin_word}:
        for i in range(len(word)):
            buckets[word[:i] + "*" + word[i + 1:]].append(word)
    q = deque([(begin_word, 1)])
    seen = {begin_word}
    while q:
        word, dist = q.popleft()
        if word == end_word:
            return dist
        for i in range(len(word)):
            for nei in buckets[word[:i] + "*" + word[i + 1:]]:
                if nei not in seen:
                    seen.add(nei)
                    q.append((nei, dist + 1))
    return 0
```

### Kth Largest Element in Array

**Difficulty:** Medium  
**Pattern:** Heap  
**Reference:** LeetCode 215  
**Solution to study:** `find_kth_largest`

**Question in plain English:** Given numbers and k, return the kth largest number.

**Like you are 12:** Keep only the k biggest numbers seen so far. The smallest inside that group is the kth largest overall.

**What the question is testing:** Whether you can turn the problem into a simple `Heap` idea and explain what each variable is remembering.

**Best approach:** Use a heap when you repeatedly need the smallest/largest item without fully sorting everything each time.

**How to recognize it:** Look for kth largest/smallest, closest points, priority scheduling, or repeated best choice.

**What to say in an interview:** Explain what the heap stores and why its top is the next useful value.

**How the solution works:**

1. Keep a min-heap containing only the largest k values seen so far.
2. The smallest value in that heap is the kth largest overall.
3. Replace the heap root only when a larger value appears.

**Beginner checks:** ties, k equals 1, k equals n.


```python
# Beginner walkthrough:
# Step 1: Keep a min-heap containing only the largest k values seen so far.
# Step 2: The smallest value in that heap is the kth largest overall.
# Step 3: Replace the heap root only when a larger value appears.

def find_kth_largest(nums, k):
    # Min-heap of size k. Time O(n log k), space O(k).
    heap = nums[:k]
    heapq.heapify(heap)
    for x in nums[k:]:
        if x > heap[0]:
            heapq.heapreplace(heap, x)
    return heap[0]
```

### Find Median from Data Stream

**Difficulty:** Hard  
**Pattern:** Two Heaps  
**Reference:** LeetCode 295  
**Solution to study:** `MedianFinder`

**Question in plain English:** Design a structure that can add numbers and return the median at any time.

**Like you are 12:** Keep the smaller half of numbers on one side and the larger half on the other. Balance them so the middle is easy to find.

**What the question is testing:** Whether you can turn the problem into a simple `Two Heaps` idea and explain what each variable is remembering.

**Best approach:** Keep the smaller half of numbers in one heap and the larger half in another so the middle is easy to read.

**How to recognize it:** Look for running median or stream problems where numbers arrive one at a time.

**What to say in an interview:** Explain the balance rule between the heaps and how the median comes from their tops.

**How the solution works:**

1. Keep the lower half in a max heap and upper half in a min heap.
2. Balance the heaps so the lower half has the same count or one extra.
3. The median is either the top of the lower heap or the average of both heap tops.

**Beginner checks:** ties, k equals 1, k equals n.


```python
# Beginner walkthrough:
# Step 1: Keep the lower half in a max heap and upper half in a min heap.
# Step 2: Balance the heaps so the lower half has the same count or one extra.
# Step 3: The median is either the top of the lower heap or the average of both heap
#         tops.

class MedianFinder:
    # Max heap for lower half, min heap for upper half. Add O(log n), find O(1).
    def __init__(self):
        self.low = []
        self.high = []

    def addNum(self, num):
        heapq.heappush(self.low, -num)
        heapq.heappush(self.high, -heapq.heappop(self.low))
        if len(self.high) > len(self.low):
            heapq.heappush(self.low, -heapq.heappop(self.high))

    def findMedian(self):
        if len(self.low) > len(self.high):
            return -self.low[0]
        return (-self.low[0] + self.high[0]) / 2
```

### Task Scheduler

**Difficulty:** Medium  
**Pattern:** Heap + Greedy  
**Reference:** LeetCode 621  
**Solution to study:** `least_interval`

**Question in plain English:** Given tasks and a cooldown, return the minimum time needed to run all tasks.

**Like you are 12:** The most repeated tasks create the schedule shape. Cooldown gaps must separate copies of the same task.

**What the question is testing:** Whether you can turn the problem into a simple `Heap + Greedy` idea and explain what each variable is remembering.

**Best approach:** Use counts to reason about the busiest task first, because the most frequent item creates the tightest spacing constraint.

**How to recognize it:** Look for scheduling with cooldowns or repeated tasks that must be spaced apart.

**What to say in an interview:** Explain the frame created by the most frequent task and why idle slots may or may not be needed.

**How the solution works:**

1. The most frequent task determines the minimum frame length.
2. Arrange the most frequent task with cooldown gaps.
3. The answer is the larger of actual task count and required frame size.

**Beginner checks:** ties, k equals 1, k equals n.


```python
# Beginner walkthrough:
# Step 1: The most frequent task determines the minimum frame length.
# Step 2: Arrange the most frequent task with cooldown gaps.
# Step 3: The answer is the larger of actual task count and required frame size.

def least_interval(tasks, n):
    # Greedy formula. Time O(t), space O(1).
    counts = Counter(tasks).values()
    max_count = max(counts)
    num_max = sum(1 for c in counts if c == max_count)
    return max(len(tasks), (max_count - 1) * (n + 1) + num_max)
```

### Implement Trie

**Difficulty:** Medium  
**Pattern:** Core  
**Reference:** LeetCode 208  
**Solution to study:** `Trie`

**Question in plain English:** Build a trie that can insert words, search full words, and check prefixes.

**Like you are 12:** Store words as paths of letters. Each node says which letters can come next.

**What the question is testing:** Whether you can turn the problem into a simple `Core` idea and explain what each variable is remembering.

**Best approach:** Build a trie as a tree of characters so shared prefixes are stored once.

**How to recognize it:** Look for prefix search, autocomplete-like behavior, or many word lookups.

**What to say in an interview:** Explain that each edge is one character and `is_end` marks a complete word.

**How the solution works:**

1. Each node stores child pointers by character.
2. Insert walks or creates nodes for each character.
3. Search and prefix lookup walk the same path and check either `is_end` or existence.

**Beginner checks:** empty input, one element, duplicate values.


```python
# Beginner walkthrough:
# Step 1: Each node stores child pointers by character.
# Step 2: Insert walks or creates nodes for each character.
# Step 3: Search and prefix lookup walk the same path and check either `is_end` or
#         existence.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

class Trie:
    # Insert/search/prefix are O(word length).
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True

    def search(self, word):
        node = self._find(word)
        return bool(node and node.is_end)

    def startsWith(self, prefix):
        return self._find(prefix) is not None

    def _find(self, text):
        node = self.root
        for ch in text:
            if ch not in node.children:
                return None
            node = node.children[ch]
        return node
```

### Design Add and Search Words

**Difficulty:** Medium  
**Pattern:** Trie + DFS  
**Reference:** LeetCode 211  
**Solution to study:** `WordDictionary`

**Question in plain English:** Build a word dictionary that supports adding words and searching with . as any character.

**Like you are 12:** A dot is a wildcard, so when you see it, you may need to try every possible next letter path.

**What the question is testing:** Whether you can turn the problem into a simple `Trie + DFS` idea and explain what each variable is remembering.

**Best approach:** Store words in a trie, then use DFS when a search character can branch into several possible next letters.

**How to recognize it:** Look for wildcard word search or prefix matching with branching choices.

**What to say in an interview:** Explain that normal letters follow one path, while `.` tries every child path.

**How the solution works:**

1. Store words in a trie.
2. Normal letters follow one child.
3. A dot branches DFS into every child at that position.

**Beginner checks:** empty tree/graph, cycles, disconnected parts.


```python
# Beginner walkthrough:
# Step 1: Store words in a trie.
# Step 2: Normal letters follow one child.
# Step 3: A dot branches DFS into every child at that position.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

class WordDictionary:
    # Trie + DFS for '.' wildcard.
    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word):
        node = self.root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True

    def search(self, word):
        def dfs(i, node):
            if i == len(word):
                return node.is_end
            ch = word[i]
            if ch == ".":
                return any(dfs(i + 1, child) for child in node.children.values())
            return ch in node.children and dfs(i + 1, node.children[ch])
        return dfs(0, self.root)
```

### Word Search II

**Difficulty:** Hard  
**Pattern:** Trie + Backtrack  
**Reference:** LeetCode 212  
**Solution to study:** `find_words`

**Question in plain English:** Given a board and words, return all words that can be formed by walking through neighboring cells.

**Like you are 12:** Put all words in a trie, then walk the board. Stop early when the current path is not a prefix of any word.

**What the question is testing:** Whether you can turn the problem into a simple `Trie + Backtrack` idea and explain what each variable is remembering.

**Best approach:** Use a trie to know which prefixes are worth exploring, and backtracking to try board paths without reusing a cell.

**How to recognize it:** Look for word search over a grid with many target words.

**What to say in an interview:** Explain that the trie prunes paths that cannot become any word.

**How the solution works:**

1. Build a trie from all words.
2. Backtrack from each board cell while following trie edges.
3. When a trie node has a full word, add it and clear it to avoid duplicates.

**Beginner checks:** prefix that is not a word, duplicate words, dot wildcard.


```python
# Beginner walkthrough:
# Step 1: Build a trie from all words.
# Step 2: Backtrack from each board cell while following trie edges.
# Step 3: When a trie node has a full word, add it and clear it to avoid duplicates.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None
        self.is_end = False

def find_words(board, words):
    # Trie prunes backtracking. Time bounded by board paths, space O(total word chars).
    root = TrieNode()
    for word in words:
        node = root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.word = word

    rows, cols = len(board), len(board[0])
    res = []

    def dfs(r, c, node):
        if r < 0 or c < 0 or r == rows or c == cols:
            return
        ch = board[r][c]
        if ch == "#" or ch not in node.children:
            return
        nxt = node.children[ch]
        if nxt.word:
            res.append(nxt.word)
            nxt.word = None
        board[r][c] = "#"
        dfs(r + 1, c, nxt)
        dfs(r - 1, c, nxt)
        dfs(r, c + 1, nxt)
        dfs(r, c - 1, nxt)
        board[r][c] = ch

    for r in range(rows):
        for c in range(cols):
            dfs(r, c, root)
    return res
```

**What to say in an interview:** Define exactly what `dp[i]` means in plain English before writing the recurrence.

```python
# Beginner walkthrough:
# Step 1: This is Fibonacci in disguise.
# Step 2: Ways to reach the current step equals previous step plus two steps back.
# Step 3: Only two previous values are needed.

def climb_stairs(n):
    # Fibonacci-style DP. Time O(n), space O(1).
    if n <= 2:
        return n
    a, b = 1, 2
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b
```

### Coin Change

**Difficulty:** Medium  
**Pattern:** 1D DP  
**Reference:** LeetCode 322  
**Solution to study:** `coin_change`

**Question in plain English:** Given coin values and an amount, return the fewest coins needed to make that amount.

**Like you are 12:** Build answers for small amounts first. For each amount, try every coin and keep the best option.

**What the question is testing:** Whether you can turn the problem into a simple `1D DP` idea and explain what each variable is remembering.

**Best approach:** Store answers to smaller versions of the problem in a one-dimensional list or a few variables.

**How to recognize it:** Look for choose/skip choices, counting ways, minimum coins, or segmenting a prefix.

**What to say in an interview:** Define exactly what `dp[i]` means in plain English before writing the recurrence.

**How the solution works:**

1. `dp[x]` stores the fewest coins needed for amount x.
2. For each amount, try every coin as the last coin used.
3. Unreachable amounts stay infinity and become `-1` at the end.

**Beginner checks:** zero amount, empty prefix, impossible state.


```python
# Beginner walkthrough:
# Step 1: `dp[x]` stores the fewest coins needed for amount x.
# Step 2: For each amount, try every coin as the last coin used.
# Step 3: Unreachable amounts stay infinity and become `-1` at the end.

def coin_change(coins, amount):
    # Unbounded knapsack min coins. Time O(amount * coins), space O(amount).
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0
    for total in range(1, amount + 1):
        for coin in coins:
            if total >= coin:
                dp[total] = min(dp[total], dp[total - coin] + 1)
    return dp[amount] if dp[amount] != float("inf") else -1
```

### Longest Increasing Subsequence

**Difficulty:** Medium  
**Pattern:** 1D DP  
**Reference:** LeetCode 300  
**Solution to study:** `length_of_lis`

**Question in plain English:** Given numbers, return the length of the longest subsequence that keeps increasing.

**Like you are 12:** Build a line of best possible endings. A smaller ending is better because it leaves more room for future numbers.

**What the question is testing:** Whether you can turn the problem into a simple `1D DP` idea and explain what each variable is remembering.

**Best approach:** Store answers to smaller versions of the problem in a one-dimensional list or a few variables.

**How to recognize it:** Look for choose/skip choices, counting ways, minimum coins, or segmenting a prefix.

**What to say in an interview:** Define exactly what `dp[i]` means in plain English before writing the recurrence.

**How the solution works:**

1. `tails[i]` stores the smallest possible tail for an increasing subsequence of length i+1.
2. Binary search finds where the current number should replace or extend a tail.
3. The length of `tails` is the LIS length.

**Beginner checks:** zero amount, empty prefix, impossible state.


```python
# Beginner walkthrough:
# Step 1: `tails[i]` stores the smallest possible tail for an increasing subsequence of
#         length i+1.
# Step 2: Binary search finds where the current number should replace or extend a tail.
# Step 3: The length of `tails` is the LIS length.

def length_of_lis(nums):
    # Patience sorting tails. Time O(n log n), space O(n).
    tails = []
    for x in nums:
        i = bisect_left(tails, x)
        if i == len(tails):
            tails.append(x)
        else:
            tails[i] = x
    return len(tails)
```

**Question in plain English:** Given a grid, count how many ways a robot can move from top-left to bottom-right using only right and down.

**Like you are 12:** Each square can be reached from the square above or the square to the left, so add those ways together.

**What the question is testing:** Whether you can turn the problem into a simple `2D DP` idea and explain what each variable is remembering.

**Best approach:** Use a table when two positions or two inputs are changing at the same time.

**How to recognize it:** Look for two strings, grid paths, or edit distance-style choices.

**What to say in an interview:** Define what `dp[i][j]` means, then explain how nearby cells build the current answer.

**How the solution works:**

1. Each cell can be reached from the top or from the left.
2. The 1D array stores the number of ways to reach each column in the current row.
3. Updating left to right adds the left value into the current column.

**Beginner checks:** empty row/string, first row/column.


```python
# Beginner walkthrough:
# Step 1: Each cell can be reached from the top or from the left.
# Step 2: The 1D array stores the number of ways to reach each column in the current
#         row.
# Step 3: Updating left to right adds the left value into the current column.

def unique_paths(m, n):
    # 2D paths optimized to 1D. Time O(mn), space O(n).
    dp = [1] * n
    for _ in range(1, m):
        for c in range(1, n):
            dp[c] += dp[c - 1]
    return dp[-1]
```

### Longest Common Subsequence

**Difficulty:** Medium  
**Pattern:** 2D DP  
**Reference:** LeetCode 1143  
**Solution to study:** `longest_common_subsequence`

**Question in plain English:** Given two strings, return the length of the longest sequence that appears in both in order.

**Like you are 12:** Compare the strings one character at a time. If letters match, extend the shared sequence. If not, keep the better previous choice.

**What the question is testing:** Whether you can turn the problem into a simple `2D DP` idea and explain what each variable is remembering.

**Best approach:** Use a table when two positions or two inputs are changing at the same time.

**How to recognize it:** Look for two strings, grid paths, or edit distance-style choices.

**What to say in an interview:** Define what `dp[i][j]` means, then explain how nearby cells build the current answer.

**How the solution works:**

1. Compare characters from both strings.
2. If they match, extend the diagonal previous result.
3. If not, keep the best result from skipping one character from either string.

**Beginner checks:** empty row/string, first row/column.


```python
# Beginner walkthrough:
# Step 1: Compare characters from both strings.
# Step 2: If they match, extend the diagonal previous result.
# Step 3: If not, keep the best result from skipping one character from either string.

def longest_common_subsequence(text1, text2):
    # 2D DP optimized to 1D. Time O(mn), space O(n).
    dp = [0] * (len(text2) + 1)
    for a in text1:
        prev = 0
        for j, b in enumerate(text2, 1):
            temp = dp[j]
            if a == b:
                dp[j] = prev + 1
            else:
                dp[j] = max(dp[j], dp[j - 1])
            prev = temp
    return dp[-1]
```

### Word Break

**Difficulty:** Medium  
**Pattern:** 1D DP  
**Reference:** LeetCode 139  
**Solution to study:** `word_break`

**Question in plain English:** Given a string and dictionary, return whether the string can be split into dictionary words.

**Like you are 12:** Mark positions in the string that can be reached by using valid words. If the end is reachable, it works.

**What the question is testing:** Whether you can turn the problem into a simple `1D DP` idea and explain what each variable is remembering.

**Best approach:** Store answers to smaller versions of the problem in a one-dimensional list or a few variables.

**How to recognize it:** Look for choose/skip choices, counting ways, minimum coins, or segmenting a prefix.

**What to say in an interview:** Define exactly what `dp[i]` means in plain English before writing the recurrence.

**How the solution works:**

1. `dp[i]` means the prefix `s[:i]` can be segmented.
2. Try every split point `j` before `i`.
3. If `dp[j]` is true and `s[j:i]` is a word, then `dp[i]` is true.

**Beginner checks:** zero amount, empty prefix, impossible state.


```python
# Beginner walkthrough:
# Step 1: `dp[i]` means the prefix `s[:i]` can be segmented.
# Step 2: Try every split point `j` before `i`.
# Step 3: If `dp[j]` is true and `s[j:i]` is a word, then `dp[i]` is true.

def word_break(s, word_dict):
    # dp[i] means s[:i] can be segmented. Time O(n^2), space O(n).
    words = set(word_dict)
    dp = [False] * (len(s) + 1)
    dp[0] = True
    for i in range(1, len(s) + 1):
        for j in range(i):
            if dp[j] and s[j:i] in words:
                dp[i] = True
                break
    return dp[-1]
```

### Edit Distance

**Difficulty:** Hard  
**Pattern:** 2D DP  
**Reference:** LeetCode 72  
**Solution to study:** `min_distance`

**Question in plain English:** Given two words, return the fewest insert, delete, or replace operations needed to change one into the other.

**Like you are 12:** Compare prefixes of the two words. If the last letters differ, try the three allowed edits and choose the cheapest.

**What the question is testing:** Whether you can turn the problem into a simple `2D DP` idea and explain what each variable is remembering.

**Best approach:** Use a table when two positions or two inputs are changing at the same time.

**How to recognize it:** Look for two strings, grid paths, or edit distance-style choices.

**What to say in an interview:** Define what `dp[i][j]` means, then explain how nearby cells build the current answer.

**How the solution works:**

1. The DP compares prefixes of both words.
2. Matching characters cost nothing beyond the diagonal state.
3. Different characters take one operation plus the best insert, delete, or replace option.

**Beginner checks:** empty row/string, first row/column.


```python
# Beginner walkthrough:
# Step 1: The DP compares prefixes of both words.
# Step 2: Matching characters cost nothing beyond the diagonal state.
# Step 3: Different characters take one operation plus the best insert, delete, or
#         replace option.

def min_distance(word1, word2):
    # Edit distance. Time O(mn), space O(n).
    prev = list(range(len(word2) + 1))
    for i, a in enumerate(word1, 1):
        curr = [i] + [0] * len(word2)
        for j, b in enumerate(word2, 1):
            if a == b:
                curr[j] = prev[j - 1]
            else:
                curr[j] = 1 + min(prev[j], curr[j - 1], prev[j - 1])
        prev = curr
    return prev[-1]
```

### Implement merge sort

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** Practice item  
**Solution to study:** `merge_sort`

**Question in plain English:** Implement merge sort, which sorts a list by splitting and merging.

**Like you are 12:** Keep splitting the list until pieces are tiny, then merge the pieces back in sorted order.

**What the question is testing:** Whether you can turn the problem into a simple `Sorting` idea and explain what each variable is remembering.

**Best approach:** Sort first when ordering makes conflicts, duplicates, or merging decisions easier.

**How to recognize it:** Look for intervals, colors/categories, or practice questions about implementing a sorting algorithm.

**What to say in an interview:** Explain how sorting changes the problem from many comparisons into a simple left-to-right pass.

**How the solution works:**

1. Split the array until each part has at most one item.
2. Recursively sort both halves.
3. Merge the two sorted halves by repeatedly taking the smaller front item.

**Beginner checks:** duplicates, already sorted, reverse sorted.


```python
# Beginner walkthrough:
# Step 1: Split the array until each part has at most one item.
# Step 2: Recursively sort both halves.
# Step 3: Merge the two sorted halves by repeatedly taking the smaller front item.

def merge_sort(arr):
    # Stable divide and conquer. Time O(n log n), space O(n).
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    return res + left[i:] + right[j:]
```

### Implement quicksort

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** Practice item  
**Solution to study:** `quick_sort`

**Question in plain English:** Implement quicksort, which sorts by choosing a pivot and partitioning values around it.

**Like you are 12:** Pick a divider number. Put smaller values on one side and bigger values on the other, then sort the sides.

**What the question is testing:** Whether you can turn the problem into a simple `Sorting` idea and explain what each variable is remembering.

**Best approach:** Sort first when ordering makes conflicts, duplicates, or merging decisions easier.

**How to recognize it:** Look for intervals, colors/categories, or practice questions about implementing a sorting algorithm.

**What to say in an interview:** Explain how sorting changes the problem from many comparisons into a simple left-to-right pass.

**How the solution works:**

1. Choose a pivot and partition the array around it.
2. Values less than or equal to pivot move before it.
3. Recursively sort the left and right partitions.

**Beginner checks:** duplicates, already sorted, reverse sorted.


```python
# Beginner walkthrough:
# Step 1: Choose a pivot and partition the array around it.
# Step 2: Values less than or equal to pivot move before it.
# Step 3: Recursively sort the left and right partitions.

def quick_sort(arr):
    # In-place quicksort. Average O(n log n), worst O(n^2), space O(log n).
    def partition(low, high):
        pivot = arr[high]
        i = low
        for j in range(low, high):
            if arr[j] <= pivot:
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
        arr[i], arr[high] = arr[high], arr[i]
        return i

    def sort(low, high):
        if low < high:
            p = partition(low, high)
            sort(low, p - 1)
            sort(p + 1, high)

    sort(0, len(arr) - 1)
    return arr
```

### Sort Colors

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** LeetCode 75  
**Solution to study:** `sort_colors`

**Question in plain English:** Given 0s, 1s, and 2s, sort them in-place.

**Like you are 12:** Use three zones: red/0s on the left, blue/2s on the right, and unknown values in the middle.

**What the question is testing:** Whether you can turn the problem into a simple `Sorting` idea and explain what each variable is remembering.

**Best approach:** Sort first when ordering makes conflicts, duplicates, or merging decisions easier.

**How to recognize it:** Look for intervals, colors/categories, or practice questions about implementing a sorting algorithm.

**What to say in an interview:** Explain how sorting changes the problem from many comparisons into a simple left-to-right pass.

**How the solution works:**

1. Maintain three regions: zeros on the left, twos on the right, unknown in the middle.
2. `mid` scans the unknown region.
3. Swap zeros left and twos right until all values are placed.

**Beginner checks:** duplicates, already sorted, reverse sorted.


```python
# Beginner walkthrough:
# Step 1: Maintain three regions: zeros on the left, twos on the right, unknown in the
#         middle.
# Step 2: `mid` scans the unknown region.
# Step 3: Swap zeros left and twos right until all values are placed.

def sort_colors(nums):
    # Dutch National Flag. Time O(n), space O(1).
    low = mid = 0
    high = len(nums) - 1
    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
```

### Merge Intervals

**Difficulty:** Practice  
**Pattern:** Sorting  
**Reference:** LeetCode 56  
**Solution to study:** `merge_intervals`

**Question in plain English:** Given intervals, merge all overlapping intervals.

**Like you are 12:** Sort intervals by start time. If the next interval touches the current one, combine them; otherwise start a new interval.

**What the question is testing:** Whether you can turn the problem into a simple `Sorting` idea and explain what each variable is remembering.

**Best approach:** Sort first when ordering makes conflicts, duplicates, or merging decisions easier.

**How to recognize it:** Look for intervals, colors/categories, or practice questions about implementing a sorting algorithm.

**What to say in an interview:** Explain how sorting changes the problem from many comparisons into a simple left-to-right pass.

**How the solution works:**

1. Sort intervals by start time.
2. If the next interval starts after the previous end, start a new interval.
3. Otherwise, extend the previous end if needed.

**Beginner checks:** duplicates, already sorted, reverse sorted.


```python
# Beginner walkthrough:
# Step 1: Sort intervals by start time.
# Step 2: If the next interval starts after the previous end, start a new interval.
# Step 3: Otherwise, extend the previous end if needed.

def merge_intervals(intervals):
    # Sort by start, merge overlaps. Time O(n log n), space O(n).
    intervals.sort()
    merged = []
    for start, end in intervals:
        if not merged or start > merged[-1][1]:
            merged.append([start, end])
        else:
            merged[-1][1] = max(merged[-1][1], end)
    return merged
```
